### Лекция: Функции активации в машинном обучении

Функции активации играют ключевую роль в моделях машинного обучения, особенно в нейронных сетях. Они определяют выход нейрона и помогают моделям решать сложные задачи путем нелинейного преобразования входных данных. В этой лекции мы рассмотрим несколько популярных функций активации, их формулы, производные и свойства.

#### 1. Линейная (или идентичная) функция активации

Линейная функция активации — это самая простая функция активации, которая просто возвращает входное значение.

**Формула:**
$$ f(x) = x $$

**Производная:**
$$ f'(x) = 1 $$

**Свойства:**
- Простая реализация.
- Не вводит нелинейности, поэтому не подходит для глубоких сетей.
- Используется в выходных слоях регрессии.

#### 2. Сигмоидная функция

Сигмоидная функция активации сжимает входное значение в диапазон от 0 до 1.

**Формула:**
$$ f(x) = \frac{1}{1 + e^{-x}} $$

**Производная:**
$$ f'(x) = f(x) \cdot (1 - f(x)) $$

**Свойства:**
- Вводит нелинейность.
- Проблемы с затуханием градиента, особенно при больших или малых значениях $ x $.
- Используется в задачах бинарной классификации.

#### 3. Гиперболический тангенс (tanh)

Функция гиперболического тангенса преобразует входные значения в диапазон от -1 до 1.

**Формула:**
$$ f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

**Производная:**
$$ f'(x) = 1 - f(x)^2 $$

**Свойства:**
- Вводит нелинейность.
- Более широкая динамика значений по сравнению с сигмоидой (от -1 до 1).
- Проблемы с затуханием градиента, но менее выражены по сравнению с сигмоидой.

#### 4. Rectified Linear Unit (ReLU)

ReLU — одна из самых популярных функций активации, особенно в глубоких нейронных сетях.

**Формула:**
$$ f(x) = \max(0, x) $$

**Производная:**
$$ f'(x) = \begin{cases}
0, & \text{если } x \leq 0 \\
1, & \text{если } x > 0
\end{cases} $$

**Свойства:**
- Вводит нелинейность.
- Способствует ускорению обучения благодаря ненасыщающейся природе.
- Может возникнуть проблема "мертвых" нейронов, когда нейрон никогда не активируется.

#### 5. Leaky ReLU

Leaky ReLU является модификацией ReLU и решает проблему "мертвых" нейронов.

**Формула:**
$$ f(x) = \begin{cases}
\alpha x, & \text{если } x \leq 0 \\
x, & \text{если } x > 0
\end{cases} $$

где $ \alpha $ — небольшая положительная константа (например, 0.01).

**Производная:**
$$ f'(x) = \begin{cases}
\alpha, & \text{если } x \leq 0 \\
1, & \text{если } x > 0
\end{cases} $$

**Свойства:**
- Вводит нелинейность.
- Предотвращает проблему "мертвых" нейронов.
- Немного увеличивает вычислительную сложность по сравнению с ReLU.

#### 6. Exponential Linear Unit (ELU)

ELU — это функция активации, которая также решает проблему "мертвых" нейронов и стабилизирует средние значения активаций ближе к нулю.

**Формула:**
$$ f(x) = \begin{cases}
\alpha (e^x - 1), & \text{если } x \leq 0 \\
x, & \text{если } x > 0
\end{cases} $$

где $ \alpha $ — положительная константа.

**Производная:**
$$ f'(x) = \begin{cases}
\alpha e^x, & \text{если } x \leq 0 \\
1, & \text{если } x > 0
\end{cases} $$

**Свойства:**
- Вводит нелинейность.
- Предотвращает проблему "мертвых" нейронов.
- Стабилизирует обучение, поддерживая активации ближе к нулю.

#### 7. Swish

Swish — это новая функция активации, предложенная исследователями от Google. Она показала, что может превосходить ReLU и его варианты в некоторых задачах.

**Формула:**
$$ f(x) = x \cdot \sigma(x) $$
где $ \sigma(x) $ — сигмоидная функция активации.

**Производная:**
$$ f'(x) = f(x) + \sigma(x) \cdot (1 - f(x)) $$

**Свойства:**
- Вводит нелинейность.
- Избегает проблемы "мертвых" нейронов.
- Мягко и непрерывно включает отрицательные значения, что может улучшить обучение.

#### 8. Softmax

Softmax обычно используется в выходных слоях нейронных сетей для многоклассовой классификации. Эта функция активации преобразует вектор значений в вероятности, сумма которых равна 1.

**Формула:**
$$ f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}} $$

**Производная:**
Производная Softmax по $ x_i $ для $ i = j $:
$$ \frac{\partial f(x_i)}{\partial x_i} = f(x_i) \cdot (1 - f(x_i)) $$

Производная Softmax по $ x_i $ для $ i \neq j $:
$$ \frac{\partial f(x_i)}{\partial x_j} = -f(x_i) \cdot f(x_j) $$

**Свойства:**
- Преобразует выходные значения в вероятности.
- Используется в задачах многоклассовой классификации.
- Гарантирует, что сумма всех выходных значений равна 1.

### Заключение

Каждая функция активации имеет свои преимущества и недостатки, и выбор функции активации зависит от конкретной задачи и архитектуры модели. Важно понимать свойства и поведение каждой функции активации, чтобы сделать осознанный выбор и достичь наилучших результатов в обучении моделей машинного обучения.



Для построения графиков функций активации и их производных на Python мы будем использовать библиотеки Matplotlib и NumPy. Мы создадим функции для каждой активации, а затем нарисуем их графики.

Код на Python для построения графиков

```python
import numpy as np
import matplotlib.pyplot as plt

# Функции активации и их производные
def linear(x):
    return x

def d_linear(x):
    return np.ones_like(x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def d_leaky_relu(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def d_elu(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

def swish(x):
    return x * sigmoid(x)

def d_swish(x):
    s = sigmoid(x)
    return swish(x) + s * (1 - swish(x))

# Диапазон значений для графиков
x = np.linspace(-10, 10, 400)

# Построение графиков
fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(12, 24))
plt.subplots_adjust(hspace=0.4)

# Линейная функция
axes[0, 0].plot(x, linear(x))
axes[0, 0].set_title('Linear Activation Function')
axes[0, 0].axhline(y=0, color='k', linestyle='-')
axes[0, 0].axvline(x=0, color='k', linestyle='-')
axes[0, 1].plot(x, d_linear(x))
axes[0, 1].set_title('Derivative of Linear Activation Function')
axes[0, 1].axhline(y=0, color='k', linestyle='-')
axes[0, 1].axvline(x=0, color='k', linestyle='-')

# Сигмоида
axes[1, 0].plot(x, sigmoid(x))
axes[1, 0].set_title('Sigmoid Activation Function')
axes[1, 0].axhline(y=0, color='k', linestyle='-')
axes[1, 0].axvline(x=0, color='k', linestyle='-')
axes[1, 1].plot(x, d_sigmoid(x))
axes[1, 1].set_title('Derivative of Sigmoid Activation Function')
axes[1, 1].axhline(y=0, color='k', linestyle='-')
axes[1, 1].axvline(x=0, color='k', linestyle='-')

# Гиперболический тангенс
axes[2, 0].plot(x, tanh(x))
axes[2, 0].set_title('Tanh Activation Function')
axes[2, 0].axhline(y=0, color='k', linestyle='-')
axes[2, 0].axvline(x=0, color='k', linestyle='-')
axes[2, 1].plot(x, d_tanh(x))
axes[2, 1].set_title('Derivative of Tanh Activation Function')
axes[2, 1].axhline(y=0, color='k', linestyle='-')
axes[2, 1].axvline(x=0, color='k', linestyle='-')

# ReLU
axes[3, 0].plot(x, relu(x))
axes[3, 0].set_title('ReLU Activation Function')
axes[3, 0].axhline(y=0, color='k', linestyle='-')
axes[3, 0].axvline(x=0, color='k', linestyle='-')
axes[3, 1].plot(x, d_relu(x))
axes[3, 1].set_title('Derivative of ReLU Activation Function')
axes[3, 1].axhline(y=0, color='k', linestyle='-')
axes[3, 1].axvline(x=0, color='k', linestyle='-')

# Leaky ReLU
axes[4, 0].plot(x, leaky_relu(x))
axes[4, 0].set_title('Leaky ReLU Activation Function')
axes[4, 0].axhline(y=0, color='k', linestyle='-')
axes[4, 0].axvline(x=0, color='k', linestyle='-')
axes[4, 1].plot(x, d_leaky_relu(x))
axes[4, 1].set_title('Derivative of Leaky ReLU Activation Function')
axes[4, 1].axhline(y=0, color='k', linestyle='-')
axes[4, 1].axvline(x=0, color='k', linestyle='-')

# ELU
axes[5, 0].plot(x, elu(x))
axes[5, 0].set_title('ELU Activation Function')
axes[5, 0].axhline(y=0, color='k', linestyle='-')
axes[5, 0].axvline(x=0, color='k', linestyle='-')
axes[5, 1].plot(x, d_elu(x))
axes[5, 1].set_title('Derivative of ELU Activation Function')
axes[5, 1].axhline(y=0, color='k', linestyle='-')
axes[5, 1].axvline(x=0, color='k', linestyle='-')

# Swish
axes[6, 0].plot(x, swish(x))
axes[6, 0].set_title('Swish Activation Function')
axes[6, 0].axhline(y=0, color='k', linestyle='-')
axes[6, 0].axvline(x=0, color='k', linestyle='-')
axes[6, 1].plot(x, d_swish(x))
axes[6, 1].set_title('Derivative of Swish Activation Function')
axes[6, 1].axhline(y=0, color='k', linestyle='-')
axes[6, 1].axvline(x=0, color='k', linestyle='-')

# Softmax
softmax = lambda x: np.exp(x) / np.sum(np.exp(x), axis=0)
d_softmax = lambda x: softmax(x) * (1 - softmax(x)) # Примерная производная
x_softmax = np.linspace(-2, 2, 400)

axes[7, 0].plot(x_softmax, softmax(x_softmax))
axes[7, 0].set_title('Softmax Activation Function')
axes[7, 1].plot(x_softmax, d_softmax(x_softmax))
axes[7, 1].set_title('Derivative of Softmax Activation Function')
axes[7, 1].axhline(y=0, color='k', linestyle='-')
axes[7, 1].axvline(x=0, color='k', linestyle='-')

plt.show()
```
