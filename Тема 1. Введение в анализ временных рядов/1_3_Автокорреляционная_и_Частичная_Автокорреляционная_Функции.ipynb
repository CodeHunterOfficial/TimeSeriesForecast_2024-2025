#Автокорреляционная и Частичная Автокорреляционная Функции

#### 1. Введение

Для анализа временных рядов часто важно понимать, как значения во временном ряду зависят друг от друга на разных временных интервалах. Временные ряды могут содержать тренды, сезонные эффекты и другие скрытые закономерности, которые помогают предсказать будущие значения. ACF и PACF - это математические инструменты, которые позволяют нам исследовать и измерять взаимосвязь между значениями временного ряда на разных временных лагах.



#### 2. Определение ACF (Автокорреляционная Функция)

Автокорреляция измеряет, насколько текущее значение временного ряда зависит от его предыдущих значений. Формально автокорреляция на лаге $k$ измеряет, как сильно значения временного ряда на временных точках $t$ и $t-k$ (то есть на текущий момент и на момент, сдвинутый на $k$ шагов назад) связаны между собой.

- Если автокорреляция на лаге $k$ высока (близка к 1), это означает, что значения временного ряда в данный момент и через $k$ шагов сильно взаимосвязаны.
- Если автокорреляция на лаге $k$ низка (близка к 0), то между текущим значением и значением на лаге $k$ нет сильной связи.
- Если автокорреляция на лаге $k$ отрицательна (близка к -1), то значения временного ряда на текущий момент и на лаге $k$ имеют обратную зависимость.


##### Формальное определение ACF

ACF для временного ряда $X_t$ с лагом $k$ определяется как:

$$
\rho(k) = \frac{\text{Cov}(X_t, X_{t-k})}{\sqrt{\text{Var}(X_t) \cdot \text{Var}(X_{t-k})}}
$$

где:
- $\rho(k)$ - автокорреляция на лаге $k$,
- $\text{Cov}(X_t, X_{t-k})$ - ковариация между значениями временного ряда на расстоянии $k$ шагов,
- $\text{Var}(X_t)$ - дисперсия временного ряда.

##### Свойства ACF

1. **Симметрия**: $\rho(k) = \rho(-k)$. ACF симметрична относительно нулевого лага.
2. **Диапазон значений**: ACF находится в диапазоне от -1 до 1.
3. **Затухание**: В стационарных временных рядах автокорреляция обычно убывает с увеличением лага.

##### Интерпретация ACF

ACF помогает понять, насколько сильно текущее значение временного ряда зависит от его прошлых значений на разных интервалах времени. На графике ACF часто можно увидеть закономерности, которые указывают на:
- **Сезонность**: если автокорреляции повторяются через определенные интервалы (лаги), это указывает на наличие сезонных эффектов.
- **Тренды**: если ACF медленно убывает, это может указывать на наличие тренда.
- **Шум**: если ACF показывает низкие корреляции на всех лагах, временной ряд, вероятно, представляет собой белый шум.



#### 3. Определение PACF (Частичная Автокорреляционная Функция)

**Частичная автокорреляционная функция (PACF)** измеряет корреляцию между значениями временного ряда на конкретном лаге $k$, устраняя влияние промежуточных лагов. PACF позволяет понять непосредственную связь между текущим значением и значением на расстоянии $k$ без учета корреляций на промежуточных интервалах (лагах).

##### Формальное определение PACF

PACF для лага $k$ — это коэффициент автокорреляции, полученный после регрессии $X_t$ на $X_{t-1}, X_{t-2}, \ldots, X_{t-(k-1)}$ и $X_{t-k}$, где в модели устраняется влияние всех промежуточных лагов.

##### Свойства PACF

1. **Обнуление после определенного лага**: Для стационарных AR(п) процессов PACF обычно обнуляется на лаге $p + 1$.
2. **Диапазон значений**: PACF также находится в диапазоне от -1 до 1.

##### Интерпретация PACF

PACF помогает понять непосредственное влияние значений временного ряда на определенных лагах без опосредованных эффектов. Это полезно для определения порядка авторегрессионной модели AR(p), так как PACF показывает, сколько лагов непосредственно влияют на текущее значение.



#### 4. Применение ACF и PACF в моделировании временных рядов

ACF и PACF используются для выбора подходящих моделей временных рядов, таких как AR (авторегрессионная модель), MA (модель скользящего среднего) и ARIMA (авторегрессионная интегрированная модель скользящего среднего).

##### Применение в моделях AR(p) и MA(q)

1. **AR(p)**: В авторегрессионной модели порядка $p$ текущее значение зависит от $p$ предыдущих значений временного ряда. График PACF для такой модели показывает значительные корреляции на первых $p$ лагах и обнуление на остальных.
2. **MA(q)**: В модели скользящего среднего порядка $q$ текущее значение временного ряда зависит от $q$ случайных шумов. Для MA-модели ACF показывает значительные значения на первых $q$ лагах, а затем быстро обнуляется.
3. **ARIMA(p, d, q)**: В этой комбинированной модели ACF и PACF помогают определить значения $p$ и $q$, что важно для построения оптимальной модели.



#### 5. Пример анализа ACF и PACF

Предположим, у нас есть временной ряд, и мы хотим понять, можно ли его смоделировать как AR(p), MA(q) или ARIMA(p, d, q) процесс.

1. **Построение графиков**: Сначала строятся графики ACF и PACF.
2. **Интерпретация ACF и PACF**:
   - Если ACF убывает экспоненциально, а PACF обнуляется после первого лага, это указывает на то, что временной ряд может быть описан AR(1) моделью.
   - Если PACF убывает экспоненциально, а ACF обнуляется после второго лага, ряд может описываться MA(2) моделью.
   - Если и ACF, и PACF медленно убывают, это может указывать на необходимость использования ARIMA модели.

Таким образом, ACF и PACF — это важные инструменты для анализа временных рядов. Они позволяют исследовать структуру данных, идентифицировать тренды и сезонные эффекты, а также помогают выбрать наиболее подходящую модель для прогнозирования.

### Временной ряд
У нас есть временной ряд длиной 6:
$$
X = [10, 12, 13, 12, 11, 14]
$$

Наша цель — вычислить автокорреляцию для лагов $k = 1$, $k = 2$ и $k = 3$, а также рассчитать частичную автокорреляцию для этих лагов.

### Шаг 1: Среднее и дисперсия временного ряда

Для начала, давайте вычислим среднее и дисперсию временного ряда, так как они необходимы для вычисления автокорреляции и нормализации ковариаций.

#### 1.1 Среднее значение ($\bar{X}$)

Среднее значение временного ряда:
$$
\bar{X} = \frac{1}{6} (10 + 12 + 13 + 12 + 11 + 14) = \frac{72}{6} = 12
$$

#### 1.2 Дисперсия ($\text{Var}(X)$)

Дисперсия вычисляется по формуле:
$$
\text{Var}(X) = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2
$$

где $n$ — длина временного ряда (в нашем случае $n = 6$).

Подставим значения:
$$
\text{Var}(X) = \frac{1}{5} \left( (10 - 12)^2 + (12 - 12)^2 + (13 - 12)^2 + (12 - 12)^2 + (11 - 12)^2 + (14 - 12)^2 \right)
$$
$$
= \frac{1}{5} \left( 4 + 0 + 1 + 0 + 1 + 4 \right) = \frac{10}{5} = 2
$$

Таким образом, дисперсия $\text{Var}(X) = 2$.

### Шаг 2: Расчет ACF для лагов $k = 1, 2, 3$

#### 2.1 ACF на лаге $k = 1$

Для расчета автокорреляции на лаге 1 нужно вычислить ковариацию между значениями временного ряда и их значениями, сдвинутыми на 1 шаг назад. Ковариация для лагов $k = 1$ вычисляется по формуле:
$$
\text{Cov}(X_t, X_{t-1}) = \frac{1}{n-1} \sum_{i=2}^{n} (X_i - \bar{X})(X_{i-1} - \bar{X})
$$

Подставляем значения:
$$
\text{Cov}(X_t, X_{t-1}) = \frac{1}{5} \left( (12 - 12)(10 - 12) + (13 - 12)(12 - 12) + (12 - 12)(13 - 12) + (11 - 12)(12 - 12) + (14 - 12)(11 - 12) \right)
$$
$$
= \frac{1}{5} \left( (0)(-2) + (1)(0) + (0)(1) + (-1)(0) + (2)(-1) \right)
$$
$$
= \frac{1}{5} \left( 0 + 0 + 0 + 0 - 2 \right) = \frac{-2}{5} = -0.4
$$

Теперь вычислим ACF на лаге 1:
$$
\rho(1) = \frac{\text{Cov}(X_t, X_{t-1})}{\text{Var}(X)} = \frac{-0.4}{2} = -0.2
$$

#### 2.2 ACF на лаге $k = 2$

Для лагов $k = 2$ нужно вычислить ковариацию между значениями временного ряда и их значениями, сдвинутыми на 2 шага назад.

Ковариация для лагов $k = 2$:
$$
\text{Cov}(X_t, X_{t-2}) = \frac{1}{n-1} \sum_{i=3}^{n} (X_i - \bar{X})(X_{i-2} - \bar{X})
$$

Подставляем значения:
$$
\text{Cov}(X_t, X_{t-2}) = \frac{1}{5} \left( (13 - 12)(10 - 12) + (12 - 12)(12 - 12) + (11 - 12)(13 - 12) + (14 - 12)(12 - 12) \right)
$$
$$
= \frac{1}{5} \left( (1)(-2) + 0 + (-1)(1) + (2)(0) \right)
$$
$$
= \frac{1}{5} \left( -2 + 0 - 1 + 0 \right) = \frac{-3}{5} = -0.6
$$

Теперь вычислим ACF на лаге 2:
$$
\rho(2) = \frac{\text{Cov}(X_t, X_{t-2})}{\text{Var}(X)} = \frac{-0.6}{2} = -0.3
$$

#### 2.3 ACF на лаге $k = 3$

Для лагов $k = 3$ нужно вычислить ковариацию между значениями временного ряда и их значениями, сдвинутыми на 3 шага назад.

Ковариация для лагов $k = 3$:
$$
\text{Cov}(X_t, X_{t-3}) = \frac{1}{n-1} \sum_{i=4}^{n} (X_i - \bar{X})(X_{i-3} - \bar{X})
$$

Подставляем значения:
$$
\text{Cov}(X_t, X_{t-3}) = \frac{1}{5} \left( (12 - 12)(10 - 12) + (11 - 12)(12 - 12) + (14 - 12)(13 - 12) \right)
$$
$$
= \frac{1}{5} \left( (0)(-2) + (-1)(0) + (2)(1) \right)
$$
$$
= \frac{1}{5} \left( 0 + 0 + 2 \right) = \frac{2}{5} = 0.4
$$

Теперь вычислим ACF на лаге 3:
$$
\rho(3) = \frac{\text{Cov}(X_t, X_{t-3})}{\text{Var}(X)} = \frac{0.4}{2} = 0.2
$$

### Шаг 3: Результаты автокорреляции

Мы рассчитали значения автокорреляционной функции (ACF) для первых трех лагов:
$$
\rho(1) = -0.2, \quad \rho(2) = -0.3, \quad \rho(3) = 0.2
$$

Эти значения показывают, как значения временного ряда зависят от их предыдущих значений.

### Шаг 4: Частичная автокорреляция (PACF)

Частичная автокорреляция на лаге $k$ измеряет зависимость между текущим значением и значением на лаге $k$, исключая влияние всех промежуточных лагов. Чтобы вычислить PACF для лагов $k = 1$, $k = 2$ и $k = 3$, мы используем метод, который включает вычисление авторегрессионных моделей или прямой расчет по специальным формулам, однако для краткости и простоты здесь мы будем полагаться на результат, который можно получить с помощью соответствующих статистических пакетов, например, в Python с использованием библиотеки `statsmodels`.

Примерное поведение PACF:

- **PACF на лаге 1**: Это просто автокорреляция на лаге 1, так как других лагов нет для коррекции.
- **PACF на лаге 2**: Для лагов $k = 2$, PACF покажет зависимость, учитывая влияние лагов 1.
- **PACF на лаге 3**: Для лагов $k = 3$, PACF покажет зависимость, учитывая влияние лагов 1 и 2.



Давайте теперь пошагово разберем, как вычисляется частичная автокорреляционная функция (PACF). Для этого мы будем использовать тот же временной ряд:

$$
X = [10, 12, 13, 12, 11, 14]
$$

Мы будем вычислять PACF для первых трех лагов ($k = 1, 2, 3$).


Для вычисления частичной автокорреляционной функции (PACF) можно использовать **рекурсивные уравнения Юла-Уокера** (часто применяются для авторегрессионных процессов) и метод **регрессии**. Давайте рассмотрим оба подхода для примера временного ряда.

Рассмотрим временной ряд длиной 6:
$$
X = [10, 12, 13, 12, 11, 14]
$$

### Подход 1: Использование рекурсивной формулы Юла-Уокера

Формулы Юла-Уокера применяются для авторегрессионных процессов, и они позволяют нам последовательно вычислять частичную автокорреляцию для каждого лага. Приведем последовательность шагов и формулы для вычисления PACF для $k = 1, 2, 3$.

#### 1.1 Рекурсивная формула Юла-Уокера

Формула Юла-Уокера для частичных автокорреляций задается следующим образом:

1. **На лаге $k = 1$** частичная автокорреляция просто равна автокорреляции на этом лаге:
  $$
   \phi_{11} = \rho(1)
   $$

2. **На лаге $k > 1$**:
  $$
   \phi_{kk} = \frac{\rho(k) - \sum_{j=1}^{k-1} \phi_{k-1, j} \cdot \rho(k-j)}{1 - \sum_{j=1}^{k-1} \phi_{k-1, j} \cdot \rho(j)}
   $$
   где $\phi_{k-1, j}$ — частичные автокорреляции для предыдущего лага $k-1$.

После вычисления $\phi_{kk}$, мы также можем обновить все промежуточные частичные автокорреляции для шага $k$ с помощью формулы:
$$
\phi_{k,j} = \phi_{k-1,j} - \phi_{kk} \cdot \phi_{k-1, k-j}
$$
для $j = 1, 2, \dots, k-1$.



#### 1.2 Шаги для вычисления PACF с использованием формулы Юла-Уокера

**Шаг 1: Вычисление PACF на лаге $k = 1$**

Как мы уже рассчитали ранее, автокорреляция на лаге 1 ($\rho(1)$) равна 0.1. Следовательно:
$$
\phi_{11} = \rho(1) = 0.1
$$

**Шаг 2: Вычисление PACF на лаге $k = 2$**

Для вычисления частичной автокорреляции на лаге 2 ($\phi_{22}$) используем автокорреляции $\rho(1)$ и $\rho(2)$.

Ранее мы рассчитали:
- $\rho(1) = 0.1$
- $\rho(2) = -0.3$

Подставим в формулу для $\phi_{22}$:
$$
\phi_{22} = \frac{\rho(2) - \phi_{11} \cdot \rho(1)}{1 - \phi_{11} \cdot \rho(1)}
$$
$$
= \frac{-0.3 - (0.1) \cdot (0.1)}{1 - (0.1) \cdot (0.1)}
$$
$$
= \frac{-0.3 - 0.01}{1 - 0.01} = \frac{-0.31}{0.99} \approx -0.313
$$

Таким образом, **PACF на лаге 2**:
$$
\phi_{22} \approx -0.313
$$

**Шаг 3: Вычисление PACF на лаге $k = 3$**

Теперь используем автокорреляции $\rho(1)$, $\rho(2)$, и $\rho(3)$ для расчета PACF на лаге 3. Ранее мы рассчитали:
- $\rho(3) = 0.2$

Также нам нужны значения $\phi_{11}$ и $\phi_{22}$, которые мы уже нашли:
$$
\phi_{11} = 0.1, \quad \phi_{22} \approx -0.313
$$

Подставляем в формулу:
$$
\phi_{33} = \frac{\rho(3) - (\phi_{21} \cdot \rho(2) + \phi_{22} \cdot \rho(1))}{1 - (\phi_{21} \cdot \rho(1) + \phi_{22} \cdot \rho(2))}
$$

### Подход 2: Вычисление PACF с помощью  регрессии

Регрессия – это другой метод вычисления PACF. Она заключается в том, что на каждом шаге мы регрессируем текущее значение временного ряда $X_t$ на предыдущие значения с лагами. Остатки регрессии используются для вычисления частичной автокорреляции.



#### Шаги для пошаговой регрессии
### Шаг 1: Регрессия для PACF на лаге 1

#### Регрессия $X_t$ на $X_{t-1}$

Для вычисления **PACF на лаге 1** нам нужно регрессировать $X_t$ на $X_{t-1}$ и вычислить остатки. Однако, так как на первом лаге мы просто оцениваем прямую зависимость между $X_t$ и $X_{t-1}$, частичная автокорреляция на этом лаге будет равна обычной автокорреляции на этом лаге.

Для вычисления **ACF на лаге 1** мы уже использовали следующую формулу:

$$
\rho(1) = \frac{\sum_{t=2}^{T} (X_t - \bar{X})(X_{t-1} - \bar{X})}{\sqrt{\sum_{t=2}^{T} (X_t - \bar{X})^2 \sum_{t=2}^{T} (X_{t-1} - \bar{X})^2}}
$$

И вычислили:
$$
\rho(1) = 0.1
$$

Следовательно, **PACF на лаге 1** будет равна:
$$
\text{PACF}(1) = \rho(1) = 0.1
$$

Это просто автокорреляция на лаге 1, так как нет промежуточных лагов.



### Шаг 2: Регрессия для PACF на лаге 2

Теперь для вычисления PACF на лаге 2, мы должны регрессировать $X_t$ на **$X_{t-1}$ и $X_{t-2}$** и вычислить остатки регрессии, то есть устранить влияние лагов 1 и 2.

#### Регрессия $X_t$ на $X_{t-1}$ и $X_{t-2}$

Для вычисления PACF на лаге 2, мы проводим линейную регрессию $X_t$ на два предыдущих значения: $X_{t-1}$ и $X_{t-2}$. Математически, это выглядит как:

$$
X_t = \alpha_2 + \beta_1 X_{t-1} + \beta_2 X_{t-2} + \varepsilon_t
$$

Здесь $\varepsilon_t$ — остаток (ошибка) регрессии. После того как мы найдем $\beta_1$ и $\beta_2$, мы можем использовать остатки $\varepsilon_t$ для вычисления частичной автокорреляции на лаге 2.

##### 2.1: Пример регрессии

Предположим, что мы хотим провести регрессию для временного ряда $X = [10, 12, 13, 12, 11, 14]$. Для каждого $t$ мы строим регрессию $X_t$ на $X_{t-1}$ и $X_{t-2}$.

В нашем случае для $t = 3, 4, 5, 6$ это будет выглядеть как:

- Для $t = 3$: $X_3 = \alpha_2 + \beta_1 X_2 + \beta_2 X_1 + \varepsilon_3$
- Для $t = 4$: $X_4 = \alpha_2 + \beta_1 X_3 + \beta_2 X_2 + \varepsilon_4$
- Для $t = 5$: $X_5 = \alpha_2 + \beta_1 X_4 + \beta_2 X_3 + \varepsilon_5$
- Для $t = 6$: $X_6 = \alpha_2 + \beta_1 X_5 + \beta_2 X_4 + \varepsilon_6$

#### 2.2: Векторный вид регрессии

Для упрощения расчётов мы можем использовать векторную форму для нахождения коэффициентов $\beta_1$ и $\beta_2$. Векторный вид уравнения регрессии для временного ряда будет выглядеть так:

$$
\begin{bmatrix}
X_3 \\
X_4 \\
X_5 \\
X_6 \\
\end{bmatrix}
=
\begin{bmatrix}
X_2 & X_1 \\
X_3 & X_2 \\
X_4 & X_3 \\
X_5 & X_4 \\
\end{bmatrix}
\begin{bmatrix}
\beta_1 \\
\beta_2 \\
\end{bmatrix}
+ \begin{bmatrix}
\varepsilon_3 \\
\varepsilon_4 \\
\varepsilon_5 \\
\varepsilon_6 \\
\end{bmatrix}
$$

Здесь мы решаем систему линейных уравнений для нахождения коэффициентов $\beta_1$ и $\beta_2$.

##### 2.3: Оценка остатков и вычисление PACF

После того как мы нашли коэффициенты регрессии $\beta_1$ и $\beta_2$, вычисляем остатки $\varepsilon_t$ для каждого $t$. Это можно сделать следующим образом:

$$
\varepsilon_3 = X_3 - (\beta_1 X_2 + \beta_2 X_1)
$$
$$
\varepsilon_4 = X_4 - (\beta_1 X_3 + \beta_2 X_2)
$$
$$
\varepsilon_5 = X_5 - (\beta_1 X_4 + \beta_2 X_3)
$$
$$
\varepsilon_6 = X_6 - (\beta_1 X_5 + \beta_2 X_4)
$$

Теперь, используя остатки $\varepsilon_t$, мы вычисляем **ACF для остатков** между $X_t$ и $X_{t-2}$, что будет являться **PACF на лаге 2**.

**PACF на лаге 2** будет вычисляться как автокорреляция между остатками $\varepsilon_t$ и значениями $X_{t-2}$.



### Шаг 3: Регрессия для PACF на лаге 3

Для вычисления PACF на лаге 3, мы должны регрессировать $X_t$ на **$X_{t-1}$**, **$X_{t-2}$** и **$X_{t-3}$**. Этот процесс аналогичен тому, что мы делали для лагов 1 и 2, но теперь мы добавляем еще один лаг и будем использовать остатки от предыдущей регрессии для вычисления PACF на более высоком лаге.




```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf

# Исходный временной ряд
X = np.array([10, 12, 13, 12, 11, 14])

# Вычисление ACF и PACF
acf_values = acf(X, nlags=3)
pacf_values = pacf(X, nlags=3)

# Визуализация ACF
plt.figure(figsize=(12, 6))

# График ACF
plt.subplot(1, 2, 1)
plt.bar(range(1, 4), acf_values[1:], color='skyblue', alpha=0.7, label='ACF')
plt.axhline(0, color='black', linewidth=1)
plt.title('Автокорреляционная функция (ACF) - 3 лага')
plt.xlabel('Лаг')
plt.ylabel('ACF')
plt.xticks(range(1, 4))
plt.grid(True)

# График PACF
plt.subplot(1, 2, 2)
plt.bar(range(1, 4), pacf_values[1:], color='salmon', alpha=0.7, label='PACF')
plt.axhline(0, color='black', linewidth=1)
plt.title('Частичная автокорреляционная функция (PACF) - 3 лага')
plt.xlabel('Лаг')
plt.ylabel('PACF')
plt.xticks(range(1, 4))
plt.grid(True)

# Показать оба графика
plt.tight_layout()
plt.show()

# Вывод значений ACF и PACF
print(f"ACF для 3 лагов: {acf_values[1:]}")
print(f"PACF для 3 лагов: {pacf_values[1:]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf, pacf

# Исходный временной ряд
X = np.array([10, 12, 13, 12, 11, 14])

# Создание графиков ACF и PACF
plt.figure(figsize=(12, 6))

# График ACF
plt.subplot(1, 2, 1)
plot_acf(X, lags=3, ax=plt.gca(), alpha=0.05)  # alpha - уровень доверия (95%)
plt.title('Автокорреляционная функция (ACF) - 3 лага')

# График PACF
plt.subplot(1, 2, 2)
plot_pacf(X, lags=3, ax=plt.gca(), alpha=0.05)  # alpha - уровень доверия (95%)
plt.title('Частичная автокорреляционная функция (PACF) - 3 лага')

# Показать графики
plt.tight_layout()
plt.show()

# Вывод значений ACF и PACF
print(f"ACF для 3 лагов: {acf(X, nlags=3)[1:]}")
print(f"PACF для 3 лагов: {pacf(X, nlags=3)[1:]}")
```


### 1. Выявление тренда и сезонности с использованием ACF и PACF

Для анализа временного ряда и выявления его ключевых компонентов — тренда, сезонности и случайных колебаний — используются графики автокорреляционной функции (ACF) и частичной автокорреляционной функции (PACF). Эти графики помогают распознать структуру данных и определить подходящую модель. Рассмотрим несколько типичных ситуаций для временных рядов и разберем, как интерпретировать ACF и PACF в каждом случае.

#### Сценарий 1: **Стационарный временной ряд**
- **ACF**: значения автокорреляции быстро затухают, например, уже на первом или втором лаге.
- **PACF**: значимая корреляция на первом лаге с последующим быстрым затуханием.
- **Вывод**: ряд является стационарным и не имеет выраженного тренда или сезонности.

#### Сценарий 2: **Трендовый временной ряд**
- **ACF**: значительная автокорреляция сохраняется на многих лагах, что указывает на долгосрочную зависимость.
- **PACF**: корреляция высока на нескольких первых лагах, затем затухает.
- **Вывод**: данные содержат тренд, и для его устранения может потребоваться дифференцирование или использование модели авторегрессии (AR).

#### Сценарий 3: **Сезонный временной ряд**
- **ACF**: автокорреляция повторяется с определенной периодичностью, например, каждые 12 лагов, если наблюдается годовая сезонность.
- **PACF**: значимая корреляция может быть на нескольких первых лагах с последующим снижением.
- **Вывод**: ряд содержит сезонные колебания, и для его моделирования рекомендуется сезонная модель, например, SARIMA.



### 2. Как использовать ACF и PACF для выбора модели

Для выбора подходящей модели временного ряда, такой как ARMA, ARIMA или SARIMA, необходимо провести анализ графиков ACF и PACF. Это позволит определить порядок моделей и настроить параметры, обеспечив наиболее точное представление данных.

#### 1. **Модель AR(p)**:
   - **PACF** показывает значимость на первых \(p\) лагах, затем значения снижаются к нулю.
   - **ACF** быстро затухает.
   - **Вывод**: при наличии такой зависимости можно выбрать модель авторегрессии с \(p\)-м порядком.

#### 2. **Модель MA(q)**:
   - **ACF** показывает значимость на первых \(q\) лагах, затем значения снижаются к нулю.
   - **PACF** затухает быстро.
   - **Вывод**: в данном случае можно использовать модель скользящего среднего порядка \(q\).

#### 3. **Модель ARMA(p, q)**:
   - **ACF** и **PACF** показывают комбинацию характеристик моделей AR и MA.
   - **Вывод**: такая модель используется для стационарного ряда, когда есть зависимость и от прошлых значений, и от шумов.

#### 4. **Модель ARIMA**:
   - Если временной ряд нестационарен (например, содержит тренд), необходимо провести дифференцирование для его стационаризации.
   - После дифференцирования графики ACF и PACF помогут определить параметры \(p\) и \(q\).

#### 5. **Модель SARIMA**:
   - Если данные имеют сезонность, используются сезонные лаги для определения сезонных параметров.
   - Графики ACF и PACF помогают определить сезонные компоненты модели (например, сезонный порядок AR и MA).


### 3. Оценка остатков модели

Анализ остатков модели — важный этап при построении прогностических моделей, таких как ARIMA или SARIMA. Остатки (ошибки) представляют собой разницу между реальными и прогнозируемыми значениями временного ряда. Хорошая модель должна свести остатки к минимуму, а сами остатки должны быть случайными и не иметь значимой автокорреляции.

Использование графиков **ACF** и **PACF** для анализа остатков помогает выявить недочеты модели, такие как пропущенные сезонные или трендовые компоненты, или остаточную автокорреляцию, указывающую на необходимость внесения дополнительных параметров.

#### Ожидаемые свойства остатков:
- **Стационарность**: статистические характеристики (среднее, дисперсия) остатков не должны меняться во времени.
- **Белый шум**: остатки должны быть случайными, без значимых автокорреляций.

Если остатки показывают автокорреляцию, это может означать, что модель не учла какие-то важные зависимости (например, тренд или сезонность), и нужно доработать модель.

#### Использование ACF и PACF для анализа остатков
- **ACF остатков**:
  - Для идеальной модели график ACF остатков должен показывать белый шум, то есть случайные колебания без значимых пиков на любом лаге.
  - Если на графике ACF остатков видны значимые пики, это свидетельствует о том, что модель не захватила структуру данных на некоторых лагах.

- **PACF остатков**:
  - PACF остатков также должен показывать отсутствие значимых автокорреляций на всех лагах.
  - Если PACF показывает корреляции на дальних лагах, это говорит о недостаточности модели.



### 4. Примеры использования ACF и PACF для анализа остатков

#### Пример 1: Остатки, соответствующие белому шуму
После применения модели ARIMA остатки показывают случайные колебания и отсутствие автокорреляции.

- **ACF остатков**: не имеет значимых корреляций.
- **PACF остатков**: также не имеет значимых корреляций.

**Вывод**: модель хорошо описывает данные и может использоваться для прогнозирования.

#### Пример 2: Остатки с автокорреляцией (недостаточная модель)
После применения модели ARIMA остатки все еще показывают автокорреляцию.

- **ACF остатков**: на графике видны значимые пики на некоторых лагах.
- **PACF остатков**: значительные корреляции на нескольких лагах.

**Вывод**: модель требует доработки, например, увеличения порядка AR (параметр \(p\)) или MA (параметр \(q\)), либо добавления сезонных компонентов.

#### Пример 3: Наличие тренда или сезонности в остатках
Если остатки показывают автокорреляцию на дальних лагах, это может свидетельствовать о наличии неучтенного тренда или сезонности.

- **ACF остатков**: значимые пики на определенных периодах (например, каждые 12 лагов).
- **PACF остатков**: корреляция на дальних лагах.

**Вывод**: модель нуждается в сезонной компоненте, и возможно, следует использовать SARIMA для учета сезонности.

Примеры и визуализация белого шума и нестационарного временного ряда

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Установим seed для воспроизводимости
np.random.seed(42)

# 1. Белый шум
n = 100
white_noise = np.random.normal(loc=0, scale=1, size=n)

# 2. Нестационарный временной ряд (случайное блуждание)
random_walk = np.cumsum(np.random.normal(loc=0, scale=1, size=n))

# Создадим графики временных рядов
plt.figure(figsize=(14, 6))

plt.subplot(2, 2, 1)
plt.plot(white_noise)
plt.title("Белый шум")

plt.subplot(2, 2, 2)
plt.plot(random_walk)
plt.title("Случайное блуждание (Нестационарный ряд)")

plt.tight_layout()
plt.show()

# Построим графики ACF и PACF для белого шума
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
plot_acf(white_noise, ax=ax[0], lags=20, title="ACF: Белый шум")
plot_pacf(white_noise, ax=ax[1], lags=20, title="PACF: Белый шум")
plt.show()

# Построим графики ACF и PACF для случайного блуждания
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
plot_acf(random_walk, ax=ax[0], lags=20, title="ACF: Случайное блуждание")
plot_pacf(random_walk, ax=ax[1], lags=20, title="PACF: Случайное блуждание")
plt.show()
```
