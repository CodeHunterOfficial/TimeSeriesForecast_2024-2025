### Стационарные временные ряды

Временной ряд $y_t$ называется **стационарным**:

1. **В узком смысле**, если для любых $t_1, \ldots, t_n$ и $\tau$, вектор $(y_{t_1+\tau}, \ldots, y_{t_n+\tau})$ совпадает по распределению с вектором $(y_{t_1}, \ldots, y_{t_n})$. Это означает, что при сдвиге всех моментов времени на одно и то же число совместное распределение значений временного ряда остаётся неизменным.

2. **В широком смысле**, если выполняются следующие условия:
   - $\mathbb{E}y_t^2 < +\infty$ для любого $t$, то есть дисперсия конечна.
   - Математическое ожидание $\mathbb{E}y_t$ не зависит от $t$, то есть среднее значение временного ряда постоянно.
   - Ковариация $\text{cov}(y_{t+\tau}, y_{s+\tau}) = \text{cov}(y_t, y_s)$ для любых $t, s, \tau$. Иными словами, автокорреляция зависит только от длины временного интервала между двумя значениями.

3. Для **гауссовских распределений** (случая, когда все векторы вида $(y_{t_1}, \ldots, y_{t_n})$ имеют нормальное распределение) определения стационарности в узком и широком смысле эквивалентны. Это связано с тем, что распределение гауссовского случайного вектора полностью определяется его математическим ожиданием и ковариациями.



### **Задание**

Исследуем временной ряд $y_t = \xi_1 \cos t + \xi_2 \sin t$, где:
- $\xi_1$ и $\xi_2$ — независимые случайные величины, одинаково распределённые;
- $P(\xi_1 = 1) = P(\xi_1 = -1) = \frac{1}{2}$.

Нужно проверить:
1. Является ли этот ряд стационарным **в широком смысле**.
2. Является ли он стационарным **в узком смысле**.



### **1. Проверка стационарности в широком смысле**

#### **Шаг 1. Математическое ожидание**
Вычислим $\mathbb{E}y_t$ (математическое ожидание временного ряда):
$$
y_t = \xi_1 \cos t + \xi_2 \sin t.
$$



Математическое ожидание случайной величины $\xi$, принимающей значения $x_1, x_2, \ldots, x_n$ с соответствующими вероятностями $p_1, p_2, \ldots, p_n$, вычисляется по формуле:

$$
\mathbb{E}\xi = \sum_{i=1}^n x_i p_i.
$$

В нашем случае:
- $\xi_1$ принимает значения $1$ и $-1$ с вероятностями $P(\xi_1 = 1) = \frac{1}{2}$ и $P(\xi_1 = -1) = \frac{1}{2}$.

Подставляем в формулу для математического ожидания:
$$
\mathbb{E}\xi_1 = 1 \cdot \frac{1}{2} + (-1) \cdot \frac{1}{2}.
$$
Считаем:
$$
\mathbb{E}\xi_1 = \frac{1}{2} - \frac{1}{2} = 0.
$$

Аналогично для $\xi_2$, поскольку она имеет те же значения и распределение. Таким образом:
$$
\mathbb{E}\xi_1 = \mathbb{E}\xi_2 = 0.
$$

Это означает, что случайные величины $\xi_1$ и $\xi_2$ в среднем равны нулю.



Следовательно:
$$
\mathbb{E}y_t = \mathbb{E}(\xi_1 \cos t + \xi_2 \sin t) = \cos t \cdot \mathbb{E}\xi_1 + \sin t \cdot \mathbb{E}\xi_2 = 0.
$$

#### **Шаг 2. Ковариация**
Теперь вычислим ковариацию $\text{cov}(y_t, y_s)$. Она определяется как:
$$
\text{cov}(y_t, y_s) = \mathbb{E}(y_t y_s) - \mathbb{E}y_t \cdot \mathbb{E}y_s.
$$
Так как $\mathbb{E}y_t = 0$, то:
$$
\text{cov}(y_t, y_s) = \mathbb{E}(y_t y_s).
$$
Подставим выражение для $y_t$ и $y_s$:
$$
y_t = \xi_1 \cos t + \xi_2 \sin t, \quad y_s = \xi_1 \cos s + \xi_2 \sin s.
$$
Тогда:
$$
y_t y_s = (\xi_1 \cos t + \xi_2 \sin t)(\xi_1 \cos s + \xi_2 \sin s).
$$
Раскроем скобки:
$$
y_t y_s = \xi_1^2 \cos t \cos s + \xi_1 \xi_2 (\cos t \sin s + \sin t \cos s) + \xi_2^2 \sin t \sin s.
$$
Учитывая, что $\xi_1$ и $\xi_2$ независимы, а также $\xi_1^2 = \xi_2^2 = 1$ (так как $\xi_1, \xi_2 \in \{-1, 1\}$), получаем:
$$
\mathbb{E}y_t y_s = \mathbb{E}(\xi_1^2) \cos t \cos s + \mathbb{E}(\xi_2^2) \sin t \sin s.
$$
Так как $\mathbb{E}(\xi_1^2) = \mathbb{E}(\xi_2^2) = 1$, то:
$$
\text{cov}(y_t, y_s) = \cos t \cos s + \sin t \sin s.
$$
Используем тригонометрическую формулу:
$$
\cos t \cos s + \sin t \sin s = \cos(t - s).
$$
Итак:
$$
\text{cov}(y_t, y_s) = \cos(t - s).
$$

#### **Шаг 3. Вывод о стационарности в широком смысле**
Условия стационарности в широком смысле:
1. $\mathbb{E}y_t$ постоянно (выполнено, $\mathbb{E}y_t = 0$).
2. Ковариация $\text{cov}(y_t, y_s)$ зависит только от разницы $t - s$ (выполнено, $\text{cov}(y_t, y_s) = \cos(t - s)$).

Таким образом, ряд $y_t$ **стационарен в широком смысле**.



### **2. Проверка стационарности в узком смысле**

Для стационарности в узком смысле требуется, чтобы распределение любого вектора $(y_{t_1}, \ldots, y_{t_n})$ не зависело от сдвига времени. Рассмотрим поведение ряда при конкретных значениях $t$.

#### **Шаг 1. Распределение при $t = 0$**
При $t = 0$:
$$
y_0 = \xi_1 \cos 0 + \xi_2 \sin 0 = \xi_1.
$$
Так как $\xi_1 \in \{-1, 1\}$ и $P(\xi_1 = 1) = P(\xi_1 = -1) = \frac{1}{2}$, то:
$$
y_0 \in \{-1, 1\}.
$$

#### **Шаг 2. Распределение при $t = \pi/4$**
При $t = \pi/4$:
$$
y_{\pi/4} = \xi_1 \cos(\pi/4) + \xi_2 \sin(\pi/4).
$$
Так как $\cos(\pi/4) = \sin(\pi/4) = \frac{1}{\sqrt{2}}$, то:
$$
y_{\pi/4} = \frac{\xi_1 + \xi_2}{\sqrt{2}}.
$$
Здесь $\xi_1 + \xi_2$ принимает значения:
- $-2$, если $\xi_1 = -1$ и $\xi_2 = -1$;
- $0$, если $\xi_1 = 1$ и $\xi_2 = -1$ (или наоборот);
- $2$, если $\xi_1 = 1$ и $\xi_2 = 1$.

Вероятности:
- $P(\xi_1 + \xi_2 = -2) = P(\xi_1 = -1, \xi_2 = -1) = \frac{1}{4}$;
- $P(\xi_1 + \xi_2 = 0) = P(\xi_1 = 1, \xi_2 = -1) + P(\xi_1 = -1, \xi_2 = 1) = \frac{1}{2}$;
- $P(\xi_1 + \xi_2 = 2) = P(\xi_1 = 1, \xi_2 = 1) = \frac{1}{4}$.

Таким образом:
$$
y_{\pi/4} \in \left\{-\sqrt{2}, 0, \sqrt{2}\right\}.
$$

#### **Шаг 3. Сравнение распределений**
При $t = 0$, $y_0$ принимает значения $\{-1, 1\}$ с вероятностью $\frac{1}{2}$ для каждого.  
При $t = \pi/4$, $y_{\pi/4}$ принимает значения $\{-\sqrt{2}, 0, \sqrt{2}\}$ с вероятностями $\frac{1}{4}, \frac{1}{2}, \frac{1}{4}$.  

Так как распределения отличаются, временной ряд **не является стационарным в узком смысле**.



### **Итог**
- Временной ряд $y_t$ стационарен **в широком смысле**.
- Временной ряд **не стационарен в узком смысле**, так как распределение изменяется при сдвиге времени.



### Примеры нестационарных временных рядов


### 1. **Случайное блуждание**  
**Формула:**  
$$
y_t = y_{t-1} + \varepsilon_t,
$$
где $\varepsilon_t$ — белый шум, т.е. случайные величины с $\mathbb{E}(\varepsilon_t) = 0$ и $D(\varepsilon_t) = \sigma^2$ (не зависит от времени).

#### Математическое ожидание ($\mathbb{E}(y_t)$)
Математическое ожидание $y_t$ на каждом шаге:
$$
\mathbb{E}(y_t) = \mathbb{E}(y_{t-1} + \varepsilon_t) = \mathbb{E}(y_{t-1}) + \mathbb{E}(\varepsilon_t).
$$
Так как $\mathbb{E}(\varepsilon_t) = 0$, получаем:
$$
\mathbb{E}(y_t) = \mathbb{E}(y_{t-1}).
$$
Если начальное значение $y_0$ имеет математическое ожидание $\mathbb{E}(y_0) = \mu_0$, то:
$$
\mathbb{E}(y_t) = \mu_0 \quad \text{(постоянное значение)}.
$$
Таким образом, математическое ожидание $\mathbb{E}(y_t)$ **не меняется с течением времени**, что соответствует стационарности среднего.

#### Дисперсия ($D(y_t)$)
Теперь вычислим дисперсию. Дисперсия $y_t$ на каждом шаге:
$$
D(y_t) = D(y_{t-1} + \varepsilon_t).
$$
Так как $y_{t-1}$ и $\varepsilon_t$ независимы:
$$
D(y_t) = D(y_{t-1}) + D(\varepsilon_t).
$$
Обозначим $D(y_0) = \sigma_0^2$ — начальную дисперсию. Тогда:
$$
D(y_t) = \sigma_0^2 + t \cdot \sigma^2,
$$
где $t \cdot \sigma^2$ — накопление дисперсии с течением времени (так как каждое новое значение $\varepsilon_t$ добавляет свою дисперсию).

Таким образом, **дисперсия растет с каждым шагом**, что нарушает стационарность по дисперсии.



### 2. **Временной ряд с трендом**  
**Формула:**  
$$
y_t = \alpha + \beta t + \varepsilon_t,
$$
где $\alpha$ — константа, $\beta$ — коэффициент тренда, $\varepsilon_t$ — белый шум.

#### Математическое ожидание ($\mathbb{E}(y_t)$)
Математическое ожидание:
$$
\mathbb{E}(y_t) = \mathbb{E}(\alpha + \beta t + \varepsilon_t) = \mathbb{E}(\alpha) + \mathbb{E}(\beta t) + \mathbb{E}(\varepsilon_t).
$$
Так как $\mathbb{E}(\varepsilon_t) = 0$ и $\mathbb{E}(\alpha) = \alpha$, $\mathbb{E}(\beta t) = \beta t$, получаем:
$$
\mathbb{E}(y_t) = \alpha + \beta t.
$$
Математическое ожидание **изменяется линейно с течением времени** (за счет компоненты $\beta t$), что делает ряд нестационарным по среднему.

#### Дисперсия ($D(y_t)$)
Дисперсия:
$$
D(y_t) = D(\alpha + \beta t + \varepsilon_t).
$$
Поскольку $\alpha$ и $\beta t$ — это детерминированные величины, их дисперсии равны нулю. Остается только дисперсия белого шума:
$$
D(y_t) = D(\varepsilon_t) = \sigma^2.
$$
Таким образом, дисперсия **постоянна** и не зависит от времени, что является характерным для ряда с постоянной дисперсией.



### 3. **Временной ряд с сезонностью**  
**Формула:**  
$$
y_t = \sin(t) + \varepsilon_t,
$$
где $\sin(t)$ — периодическая функция, $\varepsilon_t$ — белый шум.

#### Математическое ожидание ($\mathbb{E}(y_t)$)
Математическое ожидание:
$$
\mathbb{E}(y_t) = \mathbb{E}(\sin(t) + \varepsilon_t) = \mathbb{E}(\sin(t)) + \mathbb{E}(\varepsilon_t).
$$
Так как $\mathbb{E}(\varepsilon_t) = 0$, получаем:
$$
\mathbb{E}(y_t) = \sin(t).
$$
Здесь математическое ожидание **периодически изменяется** со временем, так как функция $\sin(t)$ имеет периодические колебания. Следовательно, среднее значение **не является постоянным**.

#### Дисперсия ($D(y_t)$)
Дисперсия:
$$
D(y_t) = D(\sin(t) + \varepsilon_t).
$$
Поскольку $\sin(t)$ — это детерминированная функция, то её дисперсия равна нулю:
$$
D(\sin(t)) = 0.
$$
Оставшаяся дисперсия будет равна дисперсии белого шума:
$$
D(y_t) = D(\varepsilon_t) = \sigma^2.
$$
Таким образом, дисперсия **постоянна** и не зависит от времени.



### Итоговые выводы

- **Случайное блуждание**: математическое ожидание остаётся постоянным, но дисперсия растет с течением времени, что делает ряд нестационарным.
- **Временной ряд с трендом**: математическое ожидание изменяется линейно с течением времени, но дисперсия постоянна, что делает ряд нестационарным.
- **Временной ряд с сезонностью**: математическое ожидание периодически изменяется, но дисперсия постоянна, что также делает ряд нестационарным.

Нестационарность этих рядов определяется изменениями математического ожидания или дисперсии во времени.



Стационарный ряд визуально не имеет предсказуемых закономерностей. Если посмотреть на график такого ряда издалека, то он будет выглядеть горизонтальным.

Ряд можно проверить на стационарность с помощью различных статистических критериев. Наиболее популярны следующие критерии:

1. Критерий KPSS (Kwiatkowski–Phillips–Schmidt–Shin): если p-value
   $\leq 0.05$, то отвергаем гипотезу о стационарности.
2. Критерий Дики-Фуллера: если p-value
   $\leq 0.05$, то отвергаем гипотезу о **не**стационарности.

### Общий синтаксис и параметры функций `adfuller` и `kpss`

В библиотеке `statsmodels` реализованы две ключевые функции для анализа стационарности временных рядов: `adfuller` и `kpss`. Эти функции позволяют оценивать наличие единичных корней в данных, что является важным аспектом в статистическом анализе и эконометрике.

#### 1. Функция `adfuller`

**Синтаксис:**

```python
adfuller(x, maxlag=None, regression='c', autolag='AIC')
```

**Параметры:**

- **`x`**:
  - *Тип*: массив (array-like)
  - *Описание*: Временной ряд, который подлежит тестированию на стационарность. Этот параметр является обязательным.

- **`maxlag`**:
  - *Тип*: целое число или `None`
  - *Описание*: Максимальное количество лагов, которые будут учитываться в тесте. Если установлен в `None`, то количество лагов будет определено автоматически на основе данных.

- **`regression`**:
  - *Тип*: строка
  - *Описание*: Определяет тип регрессионной модели, используемой в тесте:
    - `'c'` — модель с константой (тест без тренда).
    - `'ct'` — модель с константой и трендом.
    - `'ctt'` — модель с константой, линейным трендом и квадратичным трендом.

- **`autolag`**:
  - *Тип*: строка
  - *Описание*: Метод, используемый для автоматического выбора числа лагов. Возможные значения:
    - `'AIC'` — минимизация информации Акаике.
    - `'BIC'` — минимизация байесовской информации.
    - `'t-stat'` — использование t-статистики для выбора лагов.

#### 2. Функция `kpss`

**Синтаксис:**

```python
kpss(x, regression='c', lags='legacy')
```

**Параметры:**

- **`x`**:
  - *Тип*: массив (array-like)
  - *Описание*: Временной ряд, который подлежит тестированию на стационарность. Этот параметр является обязательным.

- **`regression`**:
  - *Тип*: строка
  - *Описание*: Определяет тип регрессионной модели, используемой в тесте:
    - `'c'` — модель с константой (тест на уровень без тренда).
    - `'ct'` — модель с константой и трендом (тест на уровень с трендом).

- **`lags`**:
  - *Тип*: строка или целое число
  - *Описание*: Определяет количество лагов, используемых в тесте. Возможные значения:
    - `'legacy'` — использование фиксированного количества лагов.
    - Целое число — конкретное количество лагов, которые будут применяться в тесте.

Такми образом, функции `adfuller` и `kpss` являются важными инструментами для проверки стационарности временных рядов. Понимание их параметров и корректная настройка тестов позволяют более точно анализировать характеристики данных, что играет ключевую роль в принятии обоснованных решений в области статистического анализа и эконометрики.

````python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss

# Функция для тестирования стационарности с помощью критерия KPSS и Дики-Фуллера
def test_stationarity(series):
    # Тест Дики-Фуллера
    adf_result = adfuller(series)
    adf_statistic = adf_result[0]
    adf_p_value = adf_result[1]

    # Тест KPSS
    kpss_result = kpss(series, regression='c')
    kpss_statistic = kpss_result[0]
    kpss_p_value = kpss_result[1]

    print("Дики-Фуллер:")
    print(f"Статистика: {adf_statistic}, p-value: {adf_p_value}")
    if adf_p_value <= 0.05:
        print("Отвергаем гипотезу о нестационарности (ряд стационарен).")
    else:
        print("Не отвергаем гипотезу о нестационарности (ряд нестационарен).")

    print("\nKPSS:")
    print(f"Статистика: {kpss_statistic}, p-value: {kpss_p_value}")
    if kpss_p_value <= 0.05:
        print("Отвергаем гипотезу о стационарности (ряд нестационарен).")
    else:
        print("Не отвергаем гипотезу о стационарности (ряд стационарен).")

# Случайное блуждание (y_t = y_{t-1} + ε_t)
np.random.seed(42)
n = 100
white_noise = np.random.normal(0, 1, n)
random_walk = np.cumsum(white_noise)  # Суммируем белый шум для моделирования случайного блуждания

# Временной ряд с трендом (y_t = α + β * t + ε_t)
alpha = 5
beta = 0.2
trend_series = alpha + beta * np.arange(n) + white_noise

# Временной ряд с сезонностью (y_t = sin(t) + ε_t)
seasonal_series = np.sin(np.arange(n)) + white_noise

# Визуализация временных рядов
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(random_walk)
plt.title("Случайное блуждание")

plt.subplot(3, 1, 2)
plt.plot(trend_series)
plt.title("Временной ряд с трендом")

plt.subplot(3, 1, 3)
plt.plot(seasonal_series)
plt.title("Временной ряд с сезонностью")

plt.tight_layout()
plt.show()

# Тестируем стационарность
print("\n--- Результаты тестов для случайного блуждания ---")
test_stationarity(random_walk)

print("\n--- Результаты тестов для временного ряда с трендом ---")
test_stationarity(trend_series)

print("\n--- Результаты тестов для временного ряда с сезонностью ---")
test_stationarity(seasonal_series)
````





##**Приведение к стационарным рядам: стабилизация дисперсии**

**Зачем это нужно?**

Методы стабилизации дисперсии, такие как преобразование Бокса-Кокса, рекомендуется использовать, когда задача требует аналитики временного ряда, особенно если необходимо построить предсказательные интервалы. Если же задача заключается исключительно в построении точечного прогноза на будущее, без учета неопределенности (например, без предсказательных интервалов), то стабилизация дисперсии не является обязательной процедурой. Однако, если нас интересует не только точечный прогноз, но и оценка доверительных интервалов, то такие методы, как ARIMA, ГАРЧ (GARCH) или другие модели для прогнозирования временных рядов, обычно работают лучше с **стационарными рядами**. Стационарность ряда в данном контексте означает, что дисперсия остается постоянной во времени. Преобразование Бокса-Кокса и другие методы позволяют стабилизировать дисперсию, делая временной ряд более стационарным.

**Преобразования:**

1. **Преобразование Бокса-Кокса**  
   Это классический метод для стабилизации дисперсии. Он описывается следующим образом:
   $$
   z_t =
   \begin{cases}
   \ln(y_t), & \lambda = 0, \\
   \frac{y_t^\lambda - 1}{\lambda}, & \lambda \neq 0.
   \end{cases}
   $$
   где:
   - $y_t$ — это исходное значение временного ряда на момент времени $t$,
   - $z_t$ — преобразованное значение,
   - $\lambda$ — параметр преобразования, который определяет степень изменения данных.

   В зависимости от значения $\lambda$, преобразование может либо уменьшать дисперсию (для больших $\lambda$), либо усиливать ее (для малых $\lambda$), или же перейти к логарифмическому преобразованию, если $\lambda = 0$.

2. **Преобразование с учетом временной зависимости дисперсии**  
   Если существует предположение, что дисперсия данных $D(y_t)$ зависит от времени $t$, то можно воспользоваться другим подходом:
   $$
   z_t = \frac{y_t}{D(y_t)}
   $$
   где $D(y_t)$ — оценка дисперсии на момент времени $t$. Это преобразование помогает уменьшить влияние времени на изменчивость ряда, стабилизируя его дисперсию.

**Обратное преобразование:**

После того как построен прогноз для преобразованного ряда, необходимо выполнить обратное преобразование для возвращения прогноза к исходному масштабу. Для этого используется следующее обратное преобразование, в зависимости от значения параметра $\lambda$:

1. Если использовалось логарифмическое преобразование ($\lambda = 0$):
   $$
   y_t = \exp(z_t)
   $$

2. Для других значений $\lambda$:
   $$
   y_t = \left( z_t \cdot \lambda + 1 \right)^{1/\lambda}
   $$

Таким образом, после применения преобразования Бокса-Кокса, можно эффективно получить точечный прогноз, а также построить доверительные интервалы для будущих значений.

### Пример

Рассмотрим простой пример с временным рядом, состоящим из следующих значений продаж товара за 12 месяцев:

$$
y = [10, 12, 15, 17, 22, 29, 34, 42, 53, 60, 65, 78]
$$

1. **Оценка параметра $\lambda$:**  
   Для этого применяем метод максимального правдоподобия (или используем статистические пакеты), чтобы найти оптимальное значение $\lambda$, например, $\lambda = 0.5$.

2. **Преобразование ряда:**  
   Применяем формулу для преобразования с найденным значением $\lambda$:

   $$
   z_t = \frac{y_t^{0.5} - 1}{0.5}
   $$

3. **Построение прогноза:**  
   Строим прогноз для преобразованного ряда $z_t$, например, с помощью модели ARIMA или другого метода.

4. **Обратное преобразование:**  
   После построения прогноза для $z_t$ выполняем обратное преобразование:

   $$
   y_t = (z_t \cdot 0.5 + 1)^{1/0.5}
   $$

Этот процесс позволяет нам стабилизировать дисперсию ряда и получить более точные прогнозы.

Таким образом, применение методов стабилизации дисперсии, таких как преобразование Бокса-Кокса, помогает превратить гетероскедастичные ряды в более стационарные. Это особенно важно при построении моделей для предсказания будущих значений временных рядов и создания предсказательных интервалов. Преобразования помогают уменьшить колебания в данных и улучшить точность модели для анализа и прогнозирования.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.tsa.arima.model import ARIMA

# Исходные данные (например, месячные продажи товара)
y = np.array([10, 12, 15, 17, 22, 29, 34, 42, 53, 60, 65, 78])

# 1. Преобразование Бокса-Кокса
# Оценка параметра lambda с помощью scipy
y_transformed, lambda_ = stats.boxcox(y)

print(f"Оценка параметра λ: {lambda_}")
print("Преобразованный ряд:", y_transformed)

# 2. Строим модель ARIMA на преобразованном ряду
# Для простоты используем ARIMA(1, 0, 0)
model = ARIMA(y_transformed, order=(1, 0, 0))
model_fit = model.fit()

# Прогноз на 3 шага вперед
forecast_transformed = model_fit.forecast(steps=3)
print("Прогноз на преобразованном ряду:", forecast_transformed)

# 3. Обратное преобразование прогноза
# Если λ = 0, используем экспоненциальную функцию для логарифмического преобразования
if lambda_ != 0:
    forecast_original = (forecast_transformed * lambda_ + 1) ** (1 / lambda_)
else:
    forecast_original = np.exp(forecast_transformed)

print("Прогноз на исходном ряду (обратное преобразование):", forecast_original)

# 4. Визуализация
# Исходный ряд и прогноз
plt.plot(y, label='Исходный ряд')
plt.plot(range(len(y), len(y) + 3), forecast_original, label='Прогноз', color='red', linestyle='--')
plt.legend()
plt.show()
````

**Приведение к стационарным рядам: тренд и сезонность**

**Преобразования:**

1. **Дифференцирование ряда** — это процесс получения разности между текущим и предыдущим значением ряда. Формально, преобразование выражается как $y_t' = y_t - y_{t-1}$, где $t \in \{2, 3, \dots, T\}$. Это преобразование используется для удаления тренда из временного ряда.
   
2. **Сезонное дифференцирование ряда** — это процесс получения разности между текущим значением и значением, отстоящим на одну сезонную единицу. Формально, преобразование записывается как $y_t' = y_t - y_{t-s}$, где $t \in \{s+1, s+2, \dots, T\}$, а $s$ — длина сезона (например, $s = 12$ для месячных данных с сезонностью в год). Это преобразование используется для устранения сезонных колебаний.

Преобразования могут применяться несколько раз, в зависимости от сложности ряда. Обычно сначала применяется сезонное дифференцирование, так как оно помогает справиться с периодическими колебаниями.

**Пример:**

Предположим, мы применяем критерий KPSS для проверки стационарности ряда. Для исходного ряда p-value $< 0.01$, что свидетельствует о том, что ряд не является стационарным. После логарифмирования ряда p-value также остается $< 0.01$, что указывает на его нестационарность. Однако, после применения сезонного дифференцирования p-value становится $> 0.1$, что означает, что полученный ряд теперь можно считать стационарным. Таким образом, сезонное дифференцирование успешно устранило сезонные колебания, сделав ряд стационарным.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Пример временного ряда: продажи товара с трендом и сезонностью
np.random.seed(42)
time = np.arange(1, 121)
trend = time * 0.5
seasonality = 10 * np.sin(time / 10)
noise = np.random.normal(0, 5, len(time))
y = trend + seasonality + noise  # Генерация ряда с трендом и сезонностью

# Создадим pandas DataFrame
data = pd.DataFrame({'y': y}, index=pd.date_range('2000-01-01', periods=len(y), freq='M'))

# Функция для тестирования стационарности с помощью теста KPSS
def kpss_test(series):
    statistic, p_value, _, critical_values = kpss(series, regression='c')
    return p_value

# Тестируем стационарность исходного ряда
print("Тест для исходного ряда (p-value):", kpss_test(data['y']))

# 1. Дифференцирование ряда (снятие тренда)
data_diff = data['y'].diff().dropna()

# Тестируем стационарность после дифференцирования
print("Тест после дифференцирования (p-value):", kpss_test(data_diff))

# 2. Сезонное дифференцирование (снятие сезонности)
seasonal_period = 12  # Предположим, что сезонность составляет 12 месяцев (например, годовая сезонность)
data_seasonal_diff = data['y'].diff(seasonal_period).dropna()

# Тестируем стационарность после сезонного дифференцирования
print("Тест после сезонного дифференцирования (p-value):", kpss_test(data_seasonal_diff))

# Визуализация
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(data, label='Исходный ряд')
plt.title('Исходный ряд')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(data_diff, label='Дифференцированный ряд')
plt.title('После дифференцирования')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(data_seasonal_diff, label='Сезонно дифференцированный ряд')
plt.title('После сезонного дифференцирования')
plt.legend()

plt.tight_layout()
plt.show()
````

# Модели вида экспоненциального сглаживания

#### Простое экспоненциальное сглаживание

Когда временной ряд имеет много шума или нестабильности, это может затруднить прогнозирование и анализ. Экспоненциальное сглаживание — это метод, который помогает «сгладить» ряд, уменьшая влияние случайных колебаний и выделяя более важные тренды. Это особенно полезно, когда нам нужно улучшить прогнозирование.

Простое экспоненциальное сглаживание основано на комбинировании текущего значения ряда и его предыдущего сглаженного значения. Это позволяет сгладить данные, при этом не теряя важной информации о тренде.

Пусть $y_t$ — это наблюдаемое значение ряда на момент времени $t$. Мы строим прогноз для следующего момента времени $t+1$ по следующей формуле:

$$
\hat{y}_{t+1|t} = \alpha y_t + (1 - \alpha) \hat{y}_{t|t-1},
$$

где:

- $\hat{y}_{t+1|t}$ — это прогноз для времени $t+1$,
- $y_t$ — это текущее наблюдаемое значение,
- $\hat{y}_{t|t-1}$ — это сглаженное значение для времени $t$ (прогноз для момента времени $t$, сделанный на основе предыдущих данных),
- $\alpha$ — это коэффициент сглаживания, принимающий значения от 0 до 1.

### Интерпретация формулы

Значение $\hat{y}_{t+1|t}$ представляет собой взвешенную сумму двух компонентов:

1. **Текущее значение ряда $y_t$**, которое актуально на данный момент.
2. **Предыдущее сглаженное значение $\hat{y}_{t|t-1}$**, которое отражает тренды прошлого.

При этом вес для текущего значения $y_t$ задается коэффициентом $\alpha$, а вес для предыдущего сглаженного значения $\hat{y}_{t|t-1}$ равен $1 - \alpha$. Таким образом, более высокие значения $\alpha$ делают модель более чувствительной к последним данным, в то время как низкие значения $\alpha$ дают большее значение историческим данным.

### Свойства модели

- Когда $\alpha$ близко к 1, сглаживание происходит слабо. Прогноз будет близким к текущему значению $y_t$, а модель будет сильно реагировать на недавние изменения в ряду.
- Когда $\alpha$ близко к 0, сглаживание становится более сильным, и прогноз будет скорее усредненным значением ряда. Это означает, что тренды и колебания в данных игнорируются.
- Для поиска оптимального значения $\alpha$, которое наилучшим образом минимизирует ошибку прогноза, используется метод минимизации ошибки, например, по методу наименьших квадратов:

$$
\sum_{t=t_0}^T (\hat{y}_t(\alpha) - y_t)^2 \rightarrow \min_{\alpha}.
$$

#### Рекомендации по выбору $\alpha$:

- Если $\alpha$ в пределах от 0 до 0.3, это означает, что ряд стационарен, и сглаживание не потеряет много информации.
- Если $\alpha$ находится между 0.3 и 1, это может свидетельствовать о том, что ряд имеет тенденцию к нестабильности, и использование слишком большого $\alpha$ приведет к значительным искажениям в прогнозах.

### Примеры сглаживания

Если на графике показать исходные данные и сглаженные значения для разных значений $\alpha$, то можно будет увидеть, как сильно сглаживание уменьшает колебания данных. При большом значении $\alpha$ сглаженные данные будут близки к исходным, а при низком $\alpha$ можно будет наблюдать «запаздывание» сглаженного ряда, что особенно заметно, если в ряду присутствуют тренды или сезонность.

### Откуда взялась формула экспоненциального сглаживания?

Чтобы понять, откуда берется эта формула, можно рассмотреть задачу оптимизации прогноза с учетом экспоненциально убывающих весов. Идея заключается в том, чтобы минимизировать ошибку прогноза, при этом давая больший вес более свежим данным. Формально это можно записать как задачу минимизации ошибки по мере прогноза:

### 1. Постановка задачи: нахождение прогноза $x$
Мы хотим спрогнозировать значение $x$ для момента времени $T+1$. Для этого используем подход минимизации взвешенного среднеквадратичного отклонения (MSE) с экспоненциально убывающими весами:

$$
\sum_{t=0}^T \beta^{T-t} (y_t - x)^2 \to \min_x.
$$

Здесь:
- $y_t$ — наблюдаемые значения,
- $x$ — прогнозируемое значение,
- $\beta \in (0, 1)$ — параметр, задающий убывание веса.

Функция ошибки $\sum_{t=0}^T \beta^{T-t} (y_t - x)^2$ показывает, что более новые наблюдения $y_t$ имеют больший вес ($\beta^{T-t}$), чем старые.

### 2. Нахождение минимума
Для минимизации функции возьмем производную по $x$ и приравняем ее к нулю:

$$
\frac{\partial}{\partial x} \left( \sum_{t=0}^T \beta^{T-t} (y_t - x)^2 \right) = 0.
$$

Вычислим производную:

$$
\frac{\partial}{\partial x} \left( \sum_{t=0}^T \beta^{T-t} (y_t - x)^2 \right) = \sum_{t=0}^T \frac{\partial}{\partial x} \beta^{T-t} (y_t - x)^2.
$$

Так как производная квадратичной функции $(y_t - x)^2$ равна $2(y_t - x)(-1)$, то:

$$
\frac{\partial}{\partial x} \beta^{T-t} (y_t - x)^2 = -2 \beta^{T-t} (y_t - x).
$$

Суммируя, получаем:

$$
\sum_{t=0}^T -2 \beta^{T-t} (y_t - x) = 0.
$$

Упростим:

$$
\sum_{t=0}^T \beta^{T-t} (y_t - x) = 0.
$$

### 3. Выразим $x$
Разделим сумму на два слагаемых:

$$
\sum_{t=0}^T \beta^{T-t} y_t - x \sum_{t=0}^T \beta^{T-t} = 0.
$$

Отсюда выразим $x$:

$$
x = \frac{\sum_{t=0}^T \beta^{T-t} y_t}{\sum_{t=0}^T \beta^{T-t}}.
$$

Знаменатель $\sum_{t=0}^T \beta^{T-t}$ можно упростить. Это геометрическая прогрессия с первым членом $1$ (при $t = T$) и знаменателем $\beta$. Сумма прогрессии:

$$
\sum_{t=0}^T \beta^{T-t} = \frac{1 - \beta^{T+1}}{1 - \beta}.
$$

Для больших $T$, если $\beta < 1$, пренебрегаем $\beta^{T+1}$, так что знаменатель приближается к $\frac{1}{1 - \beta}$.

### 4. Приближение
Таким образом, $x$ можно записать как:

$$
x \approx (1 - \beta) \sum_{t=0}^T \beta^{T-t} y_t.
$$

Разложим сумму:

$$
x = (1 - \beta) \left( y_T + \beta y_{T-1} + \beta^2 y_{T-2} + \ldots \right).
$$

### 5. Итеративное представление
Если выразить $x$ для момента $T$ в рекуррентной форме, то сглаженное значение (или прогноз) для $T+1$ представляется через сглаженное значение для $T$:

$$
x_{T+1} = (1 - \beta) y_T + \beta x_T.
$$

Это и есть классическая формула экспоненциального сглаживания, где $\beta = 1 - \alpha$, а $\alpha$ задает "вес" текущего наблюдения $y_T$.

### 6. Итоговая модель
Модель экспоненциального сглаживания в общем виде записывается как:

$$
x_{T+1} = \alpha y_T + (1 - \alpha) x_T,
$$

где $\alpha \in (0, 1)$ определяет степень "забывания" старых значений.


Давайте рассмотрим конкретный пример простого экспоненциального сглаживания на временном ряду, чтобы лучше понять, как работает эта модель.

### Пример

Предположим, у нас есть временной ряд, состоящий из еженедельных продаж товара за 6 недель:

| Неделя | Продажи ($y_t$) |
|--||
| 1      | 200                 |
| 2      | 220                 |
| 3      | 250                 |
| 4      | 240                 |
| 5      | 260                 |
| 6      | 270                 |

Теперь мы хотим использовать метод простого экспоненциального сглаживания для прогнозирования продаж на 7-й неделе.

Для этого применим формулу:

$$
\hat{y}_{t+1|t} = \alpha y_t + (1 - \alpha) \hat{y}_{t|t-1},
$$

где $\hat{y}_{t+1|t}$ — это прогноз для следующей недели, $y_t$ — это фактические продажи на текущей неделе, и $\hat{y}_{t|t-1}$ — это сглаженные продажи, рассчитанные на предыдущей неделе.

Предположим, что мы выбрали значение $\alpha = 0.3$, что означает, что 30% веса отдаем текущему значению $y_t$, а остальные 70% — предыдущему сглаженному значению.

### Шаг 1: Инициализация

Для первой недели у нас нет предыдущего сглаженного значения, поэтому мы примем, что прогноз на первую неделю $\hat{y}_{1|0}$ равен фактическим продажам:

$$
\hat{y}_{1|0} = y_1 = 200.
$$

### Шаг 2: Прогнозирование для следующих недель

Теперь используем формулу для вычисления сглаженных значений и прогнозов:

#### Неделя 2:
$$
\hat{y}_{2|1} = \alpha y_1 + (1 - \alpha) \hat{y}_{1|0} = 0.3 \times 200 + 0.7 \times 200 = 200.
$$

Для недели 2 прогноз остаётся 200, так как $\hat{y}_{1|0} = 200$ (значение сглажено, и тренд не сильно выражен).

#### Неделя 3:
$$
\hat{y}_{3|2} = 0.3 \times 220 + 0.7 \times 200 = 66 + 140 = 206.
$$

#### Неделя 4:
$$
\hat{y}_{4|3} = 0.3 \times 250 + 0.7 \times 206 = 75 + 144.2 = 219.2.
$$

#### Неделя 5:
$$
\hat{y}_{5|4} = 0.3 \times 240 + 0.7 \times 219.2 = 72 + 153.44 = 225.44.
$$

#### Неделя 6:
$$
\hat{y}_{6|5} = 0.3 \times 260 + 0.7 \times 225.44 = 78 + 157.808 = 235.808.
$$

### Шаг 3: Прогноз на 7-ю неделю

Теперь мы можем сделать прогноз для 7-й недели:

$$
\hat{y}_{7|6} = 0.3 \times 270 + 0.7 \times 235.808 = 81 + 164.0656 = 245.0656.
$$

Таким образом, прогноз на 7-ю неделю будет **245.1**.

### Резюме:

- Мы использовали простое экспоненциальное сглаживание с коэффициентом сглаживания $\alpha = 0.3$.
- Прогноз для 7-й недели составил **245.1**.
- Мы видим, как сглаженное значение накапливается и смягчает колебания исходных данных, сохраняя долгосрочные тренды.

Этот метод особенно полезен, когда данные имеют шум или небольшие колебания, но при этом важен тренд, и мы не хотим, чтобы случайные колебания сильно влияли на прогноз.


````python
import numpy as np
import matplotlib.pyplot as plt

# Исходные данные
weeks = np.array([1, 2, 3, 4, 5, 6])
sales = np.array([200, 220, 250, 240, 260, 270])

# Коэффициент сглаживания
alpha = 0.3

# Массив для сглаженных значений
smoothed_values = [sales[0]]  # Начинаем с первого значения ряда

# Вычисление сглаженных значений
for t in range(1, len(sales)):
    smoothed_value = alpha * sales[t-1] + (1 - alpha) * smoothed_values[t-1]
    smoothed_values.append(smoothed_value)

# Прогноз для следующей недели (неделя 7)
forecast_week_7 = alpha * sales[-1] + (1 - alpha) * smoothed_values[-1]

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(weeks, sales, marker='o', label='Исходные продажи (y_t)', color='blue')
plt.plot(weeks, smoothed_values, marker='x', label='Сглаженные продажи', color='orange')
plt.axvline(x=7, linestyle='--', color='gray', label='Прогноз на неделю 7')
plt.scatter(7, forecast_week_7, color='red', label=f'Прогноз на 7 неделю: {forecast_week_7:.2f}', zorder=5)

# Настройки графика
plt.title('Простое экспоненциальное сглаживание')
plt.xlabel('Неделя')
plt.ylabel('Продажи')
plt.legend(loc='best')
plt.grid(True)
plt.xticks(np.arange(1, 8))
plt.tight_layout()

# Показываем график
plt.show()
````



На графике выше представлены исходные данные о продажах (синим цветом) и сглаженные значения (оранжевым цветом), полученные с использованием метода простого экспоненциального сглаживания с коэффициентом сглаживания $\alpha = 0.3$.

Мы также добавили вертикальную линию на 7-й неделе, чтобы обозначить прогноз на этот момент времени (красной точкой). Прогноз для 7-й недели составляет **245.07**.

Этот график наглядно демонстрирует, как сглаженные значения «гладят» колебания и позволяют более точно отследить долгосрочную тенденцию в данных.


##Модель Хольта

Модель Хольта используется для прогнозирования временных рядов с линейными трендами, и она может быть применена как для аддитивных, так и для мультипликативных трендов. Рассмотрим более подробно как работают её формулы и поясним шаги, которые приводят к получению коэффициентов модели.

### 1. Аддитивный линейный тренд
Модель Хольта с аддитивным линейным трендом предполагает, что прогноз на $d$ шагов вперед выражается линейной функцией от числа шагов $d$, где параметры (коэффициенты) вычисляются по формулам, аналогичным экспоненциальному сглаживанию.

Прогноз на $d$ шагов вперед для аддитивного тренда:

$$
\hat{y}_{t+d|t} = a_t + b_t \cdot d
$$

где:
- $a_t$ — уровень временного ряда в момент времени $t$,
- $b_t$ — скорость изменения уровня (или наклон тренда) в момент времени $t$.

Коэффициенты обновляются по следующим формулам:
1. Для уровня $a_t$:

$$
a_t = \alpha y_t + (1 - \alpha)(a_{t-1} + b_{t-1})
$$

где $\alpha$ — коэффициент сглаживания для уровня.

2. Для наклона $b_t$:

$$
b_t = \beta (a_t - a_{t-1}) + (1 - \beta) b_{t-1}
$$

где $\beta$ — коэффициент сглаживания для тренда.

Здесь значения $\alpha$ и $\beta$ находятся в пределах от 0 до 1, и их выбор зависит от того, насколько сильно мы хотим сгладить значения.

### 2. Мультипликативный линейный тренд
Для мультипликативного тренда модель выглядит аналогично, но с использованием произведения $a_t \cdot b_t^d$ вместо суммы.

Прогноз на $d$ шагов вперед для мультипликативного тренда:

$$
\hat{y}_{t+d|t} = a_t \cdot b_t^d
$$

Коэффициенты для мультипликативного тренда обновляются по аналогичным формулам:
1. Для уровня $a_t$:

$$
a_t = \alpha y_t + (1 - \alpha) (a_{t-1} \cdot b_{t-1})
$$

2. Для наклона $b_t$:

$$
b_t = \beta \cdot \frac{a_t}{a_{t-1}} + (1 - \beta) b_{t-1}
$$

### 3. Пояснение формирования коэффициентов
Чтобы понять, почему именно такие формулы для $a_t$ и $b_t$ являются корректными, рассмотрим шаги, которые привели к их получению.

#### Аддитивный тренд:
Для случая $d = 0$ и $d = 1$, а также для момента времени $t+1$, прогноз на следующий шаг должен быть приближенно равен уровню $a_t + b_t$. Это означает, что разность $a_{t+1} - a_t$ должна быть близка к $b_t$, чтобы значения прогнозов совпадали:

$$
a_{t+1} - a_t \approx b_t
$$

Рассмотрим ряд разностей $\Delta y_t = a_{t+1} - a_t$. Для того чтобы минимизировать ошибку в прогнозе для этих разностей, используем метод экспоненциального сглаживания:

$$
\sum_{i=0}^t (1 - \beta)^{t-i} (\Delta y_i - b)^2 \to \min_b
$$

Решением этой задачи является:

$$
b_t = \beta \Delta y_{t-1} + (1 - \beta) b_{t-1}
$$

где $\Delta y_t = a_t - a_{t-1}$. Эта формула объясняет обновление наклона $b_t$.

#### Обновление уровня $a_t$:
Для того чтобы гарантировать, что $a_{t+1} \approx b_t + a_t$, применяем экспоненциальное сглаживание к уровню временного ряда $y_t$. В качестве предыдущего сглаженного значения принимаем $a_{t-1} + b_{t-1}$, а для нового значения — $a_{t+1}$. Получаем формулу для обновления уровня:

$$
a_t = \alpha y_t + (1 - \alpha)(a_{t-1} + b_{t-1})
$$

где $\alpha$ — коэффициент сглаживания для уровня.

### 4. Итог
Модель Хольта использует методы экспоненциального сглаживания для определения двух основных параметров: уровня и тренда. Для аддитивного тренда прогноз на $d$ шагов вперед является линейной функцией, где значения обновляются с учетом предыдущих значений и коэффициентов сглаживания. В случае мультипликативного тренда прогноз строится с использованием произведения, и обновления коэффициентов происходят аналогично.

Таким образом, модель Хольта эффективно адаптируется к изменениям тренда и уровня во временных рядах, что позволяет получать точные прогнозы для линейных трендов.


Давайте рассмотрим конкретный пример использования **модели Хольта с аддитивным линейным трендом** для прогнозирования временного ряда.

### Пример:
Предположим, у нас есть следующий временной ряд, который показывает количество проданных единиц товара за 5 месяцев:

| Месяц  | $y_t$ (продажи) |
|--------|---------------------|
| 1      | 50                  |
| 2      | 55                  |
| 3      | 60                  |
| 4      | 65                  |
| 5      | 70                  |

Задача: построить прогноз на следующие 2 месяца (6-й и 7-й).

#### Шаг 1: Инициализация коэффициентов

Перед тем, как начать вычисления, нам нужно выбрать начальные значения для уровня $a_0$ и наклона $b_0$, а также коэффициенты сглаживания $\alpha$ и $\beta$. Пусть:

- $\alpha = 0.5$ (коэффициент сглаживания для уровня),
- $\beta = 0.5$ (коэффициент сглаживания для тренда).

Для начальных значений уровня и наклона примем:
- $a_0 = y_1 = 50$ (предположим, что начальный уровень равен первому наблюдению),
- $b_0 = \frac{y_2 - y_1}{1} = 55 - 50 = 5$ (наклон между первым и вторым месяцем).

#### Шаг 2: Обновление коэффициентов

Теперь будем обновлять коэффициенты $a_t$ и $b_t$ для каждого месяца.

**Для месяца 2 (t = 2):**

1. Обновляем уровень $a_2$:

$$
a_2 = \alpha y_2 + (1 - \alpha)(a_1 + b_1)
$$
$$
a_2 = 0.5 \times 55 + 0.5 \times (50 + 5) = 27.5 + 27.5 = 55
$$

2. Обновляем наклон $b_2$:

$$
b_2 = \beta (a_2 - a_1) + (1 - \beta) b_1
$$
$$
b_2 = 0.5 \times (55 - 50) + 0.5 \times 5 = 0.5 \times 5 + 2.5 = 5
$$

**Для месяца 3 (t = 3):**

1. Обновляем уровень $a_3$:

$$
a_3 = \alpha y_3 + (1 - \alpha)(a_2 + b_2)
$$
$$
a_3 = 0.5 \times 60 + 0.5 \times (55 + 5) = 30 + 30 = 60
$$

2. Обновляем наклон $b_3$:

$$
b_3 = \beta (a_3 - a_2) + (1 - \beta) b_2
$$
$$
b_3 = 0.5 \times (60 - 55) + 0.5 \times 5 = 0.5 \times 5 + 2.5 = 5
$$

**Для месяца 4 (t = 4):**

1. Обновляем уровень $a_4$:

$$
a_4 = \alpha y_4 + (1 - \alpha)(a_3 + b_3)
$$
$$
a_4 = 0.5 \times 65 + 0.5 \times (60 + 5) = 32.5 + 32.5 = 65
$$

2. Обновляем наклон $b_4$:

$$
b_4 = \beta (a_4 - a_3) + (1 - \beta) b_3
$$
$$
b_4 = 0.5 \times (65 - 60) + 0.5 \times 5 = 0.5 \times 5 + 2.5 = 5
$$

**Для месяца 5 (t = 5):**

1. Обновляем уровень $a_5$:

$$
a_5 = \alpha y_5 + (1 - \alpha)(a_4 + b_4)
$$
$$
a_5 = 0.5 \times 70 + 0.5 \times (65 + 5) = 35 + 35 = 70
$$

2. Обновляем наклон $b_5$:

$$
b_5 = \beta (a_5 - a_4) + (1 - \beta) b_4
$$
$$
b_5 = 0.5 \times (70 - 65) + 0.5 \times 5 = 0.5 \times 5 + 2.5 = 5
$$

#### Шаг 3: Прогноз на 6-й и 7-й месяц

Теперь, когда у нас есть обновленные значения уровня $a_5 = 70$ и наклона $b_5 = 5$, можем сделать прогнозы.

**Прогноз на 6-й месяц ( $d = 1$ ):**

$$
\hat{y}_{6|5} = a_5 + b_5 \cdot 1 = 70 + 5 = 75
$$

**Прогноз на 7-й месяц ( $d = 2$ ):**

$$
\hat{y}_{7|5} = a_5 + b_5 \cdot 2 = 70 + 5 \cdot 2 = 80
$$

### Результаты:

- Прогноз для 6-го месяца: $\hat{y}_{6|5} = 75$
- Прогноз для 7-го месяца: $\hat{y}_{7|5} = 80$

#### Итоговый вывод:
На основе данных о продажах за 5 месяцев, модель Хольта с аддитивным линейным трендом прогнозирует, что продажи в 6-м месяце составят 75 единиц, а в 7-м — 80 единиц.

````python
import numpy as np
import pandas as pd

# Исходные данные
data = [50, 55, 60, 65, 70]  # Пример временного ряда
alpha = 0.5  # Коэффициент сглаживания для уровня
beta = 0.5   # Коэффициент сглаживания для тренда

# Инициализация значений
a = [data[0]]  # Начальный уровень a_0
b = [data[1] - data[0]]  # Начальный наклон b_0 (разность первого и второго значения)

# Обновление коэффициентов для всех наблюдений
for t in range(1, len(data)):
    # Обновляем уровень a_t
    a_t = alpha * data[t] + (1 - alpha) * (a[t-1] + b[t-1])
    a.append(a_t)

    # Обновляем наклон b_t
    b_t = beta * (a_t - a[t-1]) + (1 - beta) * b[t-1]
    b.append(b_t)

# Прогнозы на 6-й и 7-й месяц
d = 1  # Прогноз на 1 шаг вперед (для 6-го месяца)
forecast_6 = a[-1] + b[-1] * d  # Прогноз на 6-й месяц

d = 2  # Прогноз на 2 шага вперед (для 7-го месяца)
forecast_7 = a[-1] + b[-1] * d  # Прогноз на 7-й месяц

# Вывод результатов
print("Исходные данные:")
print(data)

print("\nОбновленные коэффициенты a_t:")
print(a)

print("\nОбновленные коэффициенты b_t:")
print(b)

print(f"\nПрогноз на 6-й месяц: {forecast_6:.2f}")
print(f"Прогноз на 7-й месяц: {forecast_7:.2f}")
````


### Модель Хольта-Уинтерса

**1. Аддитивная сезонность с трендом:**

Для прогноза с аддитивной сезонностью и трендом используется следующая модель:

$$
\hat{y}_{t+d|t} = a_t + d \cdot b_t + s_{t-m+(d \mod m)}
$$

где:
- $a_t$ — уровень тренда на момент времени $t$,
- $b_t$ — скорость изменения тренда,
- $s_{t-m+(d \mod m)}$ — сезонный компонент для периода $t-m+(d \mod m)$,
- $m$ — длина сезона (количество периодов в одном сезоне).

Обновление параметров модели:

- **Уровень тренда (a_t):**
  $$
  a_t = \alpha (y_t - s_{t-m}) + (1 - \alpha) (a_{t-1} + b_{t-1})
  $$
  
- **Скорость изменения тренда (b_t):**
  $$
  b_t = \beta (a_t - a_{t-1}) + (1 - \beta) b_{t-1}
  $$

- **Сезонность (s_t):**
  $$
  s_t = \gamma (y_t - a_t) + (1 - \gamma) s_{t-m}
  $$

где:
- $\alpha, \beta, \gamma$ — параметры сглаживания для уровня, тренда и сезонности соответственно.



**2. Мультипликативная сезонность**

**Без тренда:**

Для прогноза с мультипликативной сезонностью без тренда используется следующая модель:

$$
\hat{y}_{t+d|t} = a_t \cdot s_{t-m+(d \mod m)}
$$

Обновление параметров:

- **Уровень тренда (a_t):**
  $$
  a_t = \alpha \frac{y_t}{s_{t-m}} + (1 - \alpha) a_{t-1}
  $$

- **Сезонность (s_t):**
  $$
  s_t = \gamma \frac{y_t}{a_t} + (1 - \gamma) s_{t-m}
  $$



**С линейным трендом:**

Если существует линейный тренд, то прогноз с мультипликативной сезонностью можно выразить следующей моделью:

$$
\hat{y}_{t+d|t} = (a_t + d \cdot b_t) \cdot s_{t-m+(d \mod m)}
$$

Обновление параметров:

- **Уровень тренда (a_t):**
  $$
  a_t = \alpha \frac{y_t}{s_{t-m}} + (1 - \alpha)(a_{t-1} + b_{t-1})
  $$

- **Скорость изменения тренда (b_t):**
  $$
  b_t = \beta (a_t - a_{t-1}) + (1 - \beta) b_{t-1}
  $$

- **Сезонность (s_t):**
  $$
  s_t = \gamma \frac{y_t}{a_t} + (1 - \gamma) s_{t-m}
  $$



Эти модели используют сглаживание для уровня, тренда и сезонных колебаний, что позволяет более точно прогнозировать временные ряды с различными типами сезонности и тренда.

### Пример:

Предположим, у нас есть данные о продажах продукта за несколько месяцев. Нам нужно предсказать продажи на следующий месяц. Мы выберем аддитивную модель с трендом и сезонностью.

#### Данные (продажи за 12 месяцев):
$$
y = [120, 135, 150, 145, 160, 155, 170, 185, 190, 205, 220, 210]
$$
Пусть сезонность составляет 12 месяцев, то есть $m = 12$.

#### 1. Вычислим начальные значения для параметров:

1. **Уровень тренда** $a_0$ и **скорость тренда** $b_0$:
   Пусть $a_0 = 140$, а $b_0 = 5$. Это просто начальная оценка уровня и тренда.

2. **Сезонность** $s_t$:
   Для сезона $m = 12$ нам нужно найти начальные сезонные компоненты $s_1, s_2, \dots, s_{12}$. Это можно вычислить как среднее отклонение от тренда для каждого месяца. Предположим, что для этого примера сезонность уже вычислена и равна:

$$
s = [12, 18, 16, 14, 15, 17, 14, 13, 18, 19, 21, 20]
$$

#### 2. Применим формулы для обновления параметров модели:

Предположим, что параметры сглаживания выбраны следующие:
- $\alpha = 0.1$ (для уровня тренда),
- $\beta = 0.2$ (для скорости тренда),
- $\gamma = 0.3$ (для сезонности).

Для вычисления прогноза на следующий месяц ($t = 13$) используем формулу для аддитивной модели с трендом и сезонностью:

$$
\hat{y}_{13|12} = a_{12} + b_{12} \cdot 1 + s_{13}
$$
где $s_{13}$ — это сезонный компонент для 13-го месяца, который соответствует $s_1$ (так как $m = 12$).

#### 3. Обновляем параметры:

Для обновления параметров на текущий месяц ($t = 12$) используем следующие формулы:

- **Уровень тренда (a_t):**
  $$
  a_t = \alpha (y_t - s_{t-m}) + (1 - \alpha)(a_{t-1} + b_{t-1})
  $$
  
- **Скорость изменения тренда (b_t):**
  $$
  b_t = \beta (a_t - a_{t-1}) + (1 - \beta) b_{t-1}
  $$

- **Сезонность (s_t):**
  $$
  s_t = \gamma (y_t - a_t) + (1 - \gamma) s_{t-m}
  $$

#### 4. Прогнозируем продажи на 13-й месяц:

1. Прогнозируем уровень тренда и скорость тренда для $t = 13$:

   $$
   a_{13} = a_{12} + b_{12} \cdot 1
   $$
   (в данном случае значение для $b_{12}$ и $a_{12}$ должны быть получены через итеративное обновление по предыдущим данным).

2. Сезонность для 13-го месяца $s_1$ (так как $m = 12$).

3. Прогноз для 13-го месяца:
   $$
   \hat{y}_{13|12} = a_{13} + b_{13} + s_1
   $$



Этот пример иллюстрирует, как используются сезонные и трендовые компоненты для вычисления прогноза в модели Хольта-Уинтерса. На практике все параметры ($\alpha, \beta, \gamma$) могут быть оптимизированы с использованием методов, таких как градиентный спуск или методом наименьших квадратов, чтобы минимизировать ошибку прогноза.


````python
import numpy as np
import matplotlib.pyplot as plt

# Данные (продажи за 12 месяцев)
y = np.array([120, 135, 150, 145, 160, 155, 170, 185, 190, 205, 220, 210])

# Параметры
m = 12  # Длина сезона (12 месяцев)
alpha = 0.1  # Параметр сглаживания для уровня тренда
beta = 0.2  # Параметр сглаживания для скорости тренда
gamma = 0.3  # Параметр сглаживания для сезонности

# Инициализация параметров
a = np.zeros(len(y))  # Уровень тренда
b = np.zeros(len(y))  # Скорость изменения тренда
s = np.zeros(len(y))  # Сезонность
a[0] = 140  # Начальный уровень
b[0] = 5  # Начальная скорость тренда
s[:m] = [12, 18, 16, 14, 15, 17, 14, 13, 18, 19, 21, 20]  # Начальная сезонность (примерные значения)

# Прогнозирование с использованием модели Хольта-Уинтерса с аддитивной сезонностью и трендом
forecast = np.zeros(len(y))

# Обновление параметров и прогноз
for t in range(m, len(y)):
    # Прогноз на следующий период
    forecast[t] = a[t-1] + b[t-1] + s[t-m]

    # Обновление параметров модели
    a[t] = alpha * (y[t] - s[t-m]) + (1 - alpha) * (a[t-1] + b[t-1])
    b[t] = beta * (a[t] - a[t-1]) + (1 - beta) * b[t-1]
    s[t] = gamma * (y[t] - a[t]) + (1 - gamma) * s[t-m]

# Прогноз на следующий период (месяц 13)
forecast_13 = a[-1] + b[-1] + s[0]  # Прогнозируем следующий месяц

# Визуализация
plt.plot(range(1, len(y)+1), y, label="Данные")
plt.plot(range(m+1, len(y)+1), forecast[m:], label="Прогноз", linestyle="--")
plt.title("Модель Хольта-Уинтерса с аддитивной сезонностью и трендом")
plt.xlabel("Месяцы")
plt.ylabel("Продажи")
plt.legend()
plt.show()

# Вывод прогноза на 13-й месяц
print(f"Прогноз продаж на 13-й месяц: {forecast_13:.2f}")
````

##**Адаптивное сглаживание**

При использовании стандартного экспоненциального сглаживания для временных рядов с переменным трендом, существует проблема запаздывания, особенно когда тренд меняется быстро. Стандартное сглаживание использует фиксированное значение параметра сглаживания $\alpha$, что может быть недостаточно гибким для ряда с резкими структурными изменениями. Если $\alpha$ выбрано слишком большим, модель слишком быстро реагирует на изменения, но может оставаться слишком "шумной". Если $\alpha$ выбрано маленьким, сглаживание становится слишком медленным и не успевает адаптироваться к быстрым изменениям в тренде. Адаптивное сглаживание решает эту проблему, позволяя параметру сглаживания изменяться в зависимости от текущих характеристик ряда.

### Основные идеи адаптивного экспоненциального сглаживания

Адаптивное сглаживание позволяет модели изменять скорость реакции на изменения в данных в зависимости от того, насколько значительные или внезапные эти изменения. Для этого модель оценивает ошибку прогноза и ее вариативность с течением времени. Если ошибка прогноза становится слишком большой или сильно колеблется, это сигнализирует о том, что тренд изменился, и параметр сглаживания должен быть увеличен, чтобы лучше адаптироваться к новым данным.

### Прогноз и ошибка прогноза

Пусть $y_t$ — фактическое значение ряда в момент времени $t$, а $\hat{y}_t$ — прогноз этого значения, сделанный с помощью экспоненциального сглаживания на основе предыдущих наблюдений (с шагом сглаживания $t-1$). Тогда ошибка прогноза на момент времени $t$ будет равна разнице между фактическим значением и прогнозом:
$$
\hat{\varepsilon}_t = y_t - \hat{y}_t
$$
Эта ошибка используется для оценки того, насколько хорошо модель предсказывает текущие значения.

### Определения:

1. **Среднее значение ошибки $E_t$**  
   Среднее значение ошибки $E_t$ показывает, как ошибка меняется во времени. Оно рассчитывается с использованием предыдущего значения $E_{t-1}$ и текущей ошибки:
   $$
   E_t = \gamma \hat{\varepsilon}_t + (1 - \gamma) E_{t-1}
   $$
   Где $\gamma$ — коэффициент сглаживания (обычно $\gamma$ выбирается из интервала $(0.05, 0.1)$). Этот параметр определяет, насколько сильно текущая ошибка влияет на среднее значение ошибки. Чем больше $\gamma$, тем быстрее значение $E_t$ будет адаптироваться к изменениям в ошибке.

2. **Средний разброс ошибки $A_t$**  
   Средний разброс ошибки $A_t$ позволяет оценить изменчивость ошибок прогноза. Он определяется как взвешенная сумма текущей абсолютной ошибки и предыдущего значения разброса:
   $$
   A_t = \gamma \left| \hat{\varepsilon}_t \right| + (1 - \gamma) A_{t-1}
   $$
   Разброс ошибки показывает, насколько сильно ошибка изменяется от наблюдения к наблюдению. Если $A_t$ велико, это может свидетельствовать о высоких колебаниях в ошибках прогноза, что обычно происходит, когда тренд меняется.

3. **Статистика $K_t$**  
   Статистика $K_t$ позволяет оценить, насколько адекватно модель работает в текущий момент времени. Это отношение среднего значения ошибки к среднему разбросу ошибки:
   $$
   K_t = \frac{E_t}{A_t}
   $$
   Интерпретация статистики $K_t$:
   - $K_t \approx \pm 1$ — модель систематически ошибается в одну сторону, то есть прогноз постоянно отклоняется в определенную сторону.
   - $K_t \approx 0$ — модель хорошо адаптируется к данным, не совершая значительных ошибок.

### Как работает адаптивное сглаживание?

Если модель определяет, что значение $K_t$ отличается значительно от 0, это сигнализирует о том, что изменения в ряду слишком большие, и параметры модели должны быть адаптированы. Для этого применяется метод, который регулирует значение параметра сглаживания $\alpha_t$.

1. **Корректировка параметра сглаживания**  
   Параметр сглаживания $\alpha_t$ используется для настройки чувствительности модели к изменениям данных:
   $$
   \alpha_t = \min \left( \left| K_t \right|, 1 \right)
   $$
   Здесь $\left| K_t \right|$ измеряет степень отклонения текущего прогноза от ожидаемого, и если это отклонение слишком велико, параметр $\alpha_t$ становится больше, что позволяет модели быстрее реагировать на изменения.

### Недостатки адаптивного экспоненциального сглаживания

Несмотря на свою гибкость и способность адаптироваться к изменениям, этот подход имеет несколько недостатков:

1. **Плохая реакция на одиночные выбросы**  
   Если в данных появляется одиночный выброс, модель может неправильно оценить изменение тренда и начать подстраиваться под этот выброс, что приведет к ухудшению прогноза. Это связано с тем, что адаптивное сглаживание основывается на статистике ошибок, и даже один выброс может значительно повлиять на оценки.

2. **Необходимость подбора параметра $\gamma$**  
   Параметр $\gamma$, определяющий скорость адаптации модели, требует тщательного подбора. Неправильно выбранное значение этого параметра может привести к плохим результатам — слишком большое значение сделает модель слишком чувствительной к ошибкам, а слишком маленькое — замедлит ее реакцию на изменения.



Таким образом, адаптивное сглаживание является мощным инструментом для анализа временных рядов с изменяющимся трендом. Оно позволяет быстро реагировать на изменения в данных, но при этом требует аккуратной настройки и учета возможных выбросов в ряду.

### Шаги применения адаптивного сглаживания

1. **Инициализация параметров:**
   - Пусть начальное значение $E_0 = 0$ (среднее значение ошибки на первый день).
   - Пусть начальное значение $A_0 = 0$ (средний разброс ошибки на первый день).
   - Пусть коэффициент сглаживания $\gamma = 0.1$.

Для каждого дня, начиная с 2-го, будем вычислять прогноз $\hat{y}_t$, ошибку $\hat{\varepsilon}_t$, среднее значение ошибки $E_t$, средний разброс ошибки $A_t$ и статистику $K_t$.

### Шаги вычислений

**Для дня 2:**

- Прогноз на день 1: $\hat{y}_1 = 10$
- Ошибка: $\hat{\varepsilon}_2 = y_2 - \hat{y}_1 = 12 - 10 = 2$
- Среднее значение ошибки:
  $$
  E_2 = \gamma \cdot \hat{\varepsilon}_2 + (1 - \gamma) \cdot E_1 = 0.1 \cdot 2 + 0.9 \cdot 0 = 0.2
  $$
- Средний разброс ошибки:
  $$
  A_2 = \gamma \cdot |\hat{\varepsilon}_2| + (1 - \gamma) \cdot A_1 = 0.1 \cdot 2 + 0.9 \cdot 0 = 0.2

  $$
  K_2 = \frac{E_2}{A_2} = \frac{0.2}{0.2} = 1

**Для дня 3:**

  $$
  E_3 = \gamma \cdot \hat{\varepsilon}_3 + (1 - \gamma) \cdot E_2 = 0.1 \cdot 3.8 + 0.9 \cdot 0.2 = 0.38 + 0.18 = 0.56

  $$
  A_3 = \gamma \cdot |\hat{\varepsilon}_3| + (1 - \gamma) \cdot A_2 = 0.1 \cdot 3.8 + 0.9 \cdot 0.2 = 0.38 + 0.18 = 0.56

  $$
  K_3 = \frac{E_3}{A_3} = \frac{0.56}{0.56} = 1

**Для дня 4:**

- Прогноз на день 3: $\hat{y}_3 = \hat{y}_2 + E_3 = 10.2 + 0.56 = 10.76$

  $$

- Статистика $K_4$:
  $$
  K_4 = \frac{E_4}{A_4} = \frac{0.728}{0.728} = 1

Когда данные начинают резко увеличиваться, например, на 10-й день, статистика $K_t$ может измениться. Рассмотрим данные для дня 10.

Если статистика $K_t$ продолжает оставаться значительной (например, $K_t \approx 1$), это будет сигнализировать о необходимости скорректировать параметр сглаживания $\alpha_t$, увеличив его для более быстрой адаптации модели к текущим изменениям в данных.

- В примере выше, когда произошел резкий скачок в данных, модель могла корректно подстроиться, повысив ошибки и разброс, что сигнализировало о изменении тренда.
- Модель адаптировалась к изменениям и стала более чувствительной, что позволило более точно прогнозировать продажи на будущие дни.



````python
import numpy as np
import matplotlib.pyplot as plt

# Данные о продажах
sales_data = [10, 12, 14, 13, 16, 20, 30, 40, 45, 70]

# Параметры
gamma = 0.1  # Коэффициент сглаживания
alpha = [0] * len(sales_data)  # Массив для параметра сглаживания
E = [0] * len(sales_data)  # Массив для средних ошибок
A = [0] * len(sales_data)  # Массив для среднего разброса ошибок
K = [0] * len(sales_data)  # Массив для статистики K

# Инициализация
alpha[0] = 0.1  # Начальный параметр сглаживания
E[0] = 0  # Начальная ошибка
A[0] = 0  # Начальный разброс ошибки

# Прогнозирование
for t in range(1, len(sales_data)):
    # Прогноз на основе предыдущего сглаживания
    if t == 1:
        forecast = sales_data[0]
    else:
        forecast = sales_data[t - 1] + E[t - 1]

    # Ошибка прогноза
    error = sales_data[t] - forecast

    # Средняя ошибка
    E[t] = gamma * error + (1 - gamma) * E[t - 1]

    # Средний разброс ошибки
    A[t] = gamma * abs(error) + (1 - gamma) * A[t - 1]

    # Статистика K
    K[t] = E[t] / A[t]

    # Корректировка параметра сглаживания alpha
    alpha[t] = min(abs(K[t]), 1)

# Прогнозы
forecasted_sales = [sales_data[0]]
for t in range(1, len(sales_data)):
    forecasted_sales.append(sales_data[t - 1] + E[t])

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(sales_data, label="Фактические продажи", marker='o', color='b')
plt.plot(forecasted_sales, label="Прогнозы", linestyle='--', color='r')
plt.title("Прогнозирование продаж с адаптивным экспоненциальным сглаживанием")
plt.xlabel("День")
plt.ylabel("Продажи")
plt.legend()
plt.show()

# Вывод статистики для каждого дня
for t in range(1, len(sales_data)):
    print(f"День {t}:")
    print(f"Прогноз: {forecasted_sales[t]:.2f}")
    print(f"Ошибка: {sales_data[t] - forecasted_sales[t]:.2f}")
    print(f"Средняя ошибка E_{t}: {E[t]:.2f}")
    print(f"Средний разброс A_{t}: {A[t]:.2f}")
    print(f"Статистика K_{t}: {K[t]:.2f}")
    print(f"Параметр сглаживания alpha_{t}: {alpha[t]:.2f}\n")
````
